# ਪਾਠ 01 - AI ਏਜੰਟਸ ਦਾ ਪਰਚਯ

**ਨਵਾਂ ਸਿਖਿਅਰਥੀਆਂ ਲਈ AI ਏਜੰਟਸ** ਕੋਰਸ ਦੇ ਪਹਿਲੇ ਪਾਠ ਵਿੱਚ ਤੁਹਾਡਾ ਸਵਾਗਤ ਹੈ!

ਇੱਕ **AI ਏਜੰਟ** ਇੱਕ ਪ੍ਰੋਗਰਾਮ ਹੈ ਜੋ ਵੱਡੇ ਭਾਸ਼ਾ ਮਾਡਲ (LLM) ਨੂੰ ਆਪਣੀ ਤਰਕਸ਼ੀਲ ਇੰਜਣ ਵਜੋਂ ਵਰਤਦਾ ਹੈ ਅਤੇ ਅਸਲੀ ਦੁਨੀਆ ਵਿੱਚ *ਕ੍ਰਿਆਵਾਂ* ਲੈ ਸਕਦਾ ਹੈ — API ਕਾਲ ਕਰਨਾ, ਡੇਟਾਬੇਸ ਪੁੱਛਗਿੱਛ ਕਰਨਾ ਜਾਂ ਕੋਡ ਚਲਾਉਣਾ — ਉਪਭੋਗਤਾ ਵੱਲੋਂ ਕਿਸੇ ਟੀਚੇ ਨੂੰ ਹਾਸਲ ਕਰਨ ਲਈ।

ਇਸ ਨੋਟਬੁੱਕ ਵਿੱਚ ਤੁਸੀਂ ਆਪਣਾ ਪਹਿਲਾ ਏਜੰਟ ਬਣਾਉਗੇ: ਇਕ **ਟ੍ਰੈਵਲ ਏਜੰਟ** ਜੋ ਛੁੱਟੀਆਂ ਲਈ ਮੰਜ਼ਿਲਾਂ ਦੀ ਸਿਫਾਰਸ਼ ਕਰਦਾ ਹੈ। ਰਸਤੇ ਵਿੱਚ ਤੁਸੀਂ ਸਿੱਖੋਗੇ ਕਿ ਕਿਵੇਂ:

1. **ਮਾਈਕਰੋਸੌਫਟ ਏਜੰਟ ਫ੍ਰੇਮਵਰਕ** ਦੀ ਵਰਤੋਂ ਕਰਕੇ Microsoft Foundry Agent Service ਨਾਲ ਕਨੈਕਟ ਕਰਨਾ।
2. ਏਜੰਟ ਨੂੰ ਇੱਕ **ਉਪਕਰਣ** ਦੇਣਾ — ਇੱਕ ਸਿੱਧਾ ਸਧਾਰਨ Python ਫੰਕਸ਼ਨ ਜਿਸਨੂੰ ਉਹ ਕਾਲ ਕਰ ਸਕਦਾ ਹੈ।
3. ਏਜੰਟ ਚਲਾਉਣਾ ਅਤੇ ਇਸਦੇ ਜਵਾਬ ਦੀ ਜਾਂਚ ਕਰਨੀ।
4. ਏਜੰਟ ਦੇ ਜਵਾਬ ਨੂੰ ਟੋਕਨ-ਟੋਕਨ ਸਟ੍ਰੀਮ ਕਰਨਾ।


## ਸੈਟਅਪ

ਇਸ ਨੋਟਬੁੱਕ ਨੂੰ ਚਲਾਉਣ ਤੋਂ ਪਹਿਲਾਂ, ਯਕੀਨੀ ਬਣਾਓ ਕਿ ਤੁਹਾਡੇ ਕੋਲ ਹੇਠਾਂ ਦਿੱਤੀਆਂ ਚੀਜ਼ਾਂ ਹਨ:

1. **ਇੱਕ Microsoft Foundry ਪ੍ਰੋਜੈਕਟ** ਜਿਸ ਵਿੱਚ ਇੱਕ ਡਿਪਲੋਇਡ ਚੈਟ ਮਾਡਲ (ਜਿਵੇਂ ਕਿ `gpt-5-mini`) ਹੈ।
2. **Azure CLI ਨਾਲ ਲਾਗਇਨ ਕੀਤਾ ਹੋਇਆ ਹੈ** — ਆਪਣੀ ਟਰਮੀਨਲ ਵਿੱਚ `az login` ਚਲਾਓ।
3. **ਲੋੜੀਂਦੇ ਇਨਵਾਇਰੰਮੈਂਟ ਵੈਰੀਏਬਲ ਸੈੱਟ ਕੀਤੇ ਹੋਣ:**
   - `AZURE_AI_PROJECT_ENDPOINT` — ਤੁਹਾਡੇ Microsoft Foundry ਪ੍ਰੋਜੈਕਟ ਦਾ ਐਂਡਪੋਇੰਟ।
   - `AZURE_AI_MODEL_DEPLOYMENT_NAME` — ਤੁਹਾਡੇ ਡਿਪਲੋਇਡ ਮਾਡਲ ਦਾ ਨਾਮ।

ਹੇਠਾਂ ਦਿੱਤੀ ਕੈੱਲ ਤੁਹਾਨੂੰ ਲੋੜੀਂਦੇ Python ਪੈਕੇਜਜ਼ ਇੰਸਟਾਲ ਕਰਦੀ ਹੈ।


```python
%pip install agent-framework azure-ai-projects azure-identity -q
```

```python
import logging
logging.getLogger("agent_framework.foundry").setLevel(logging.ERROR)

import os
import dotenv
from agent_framework.foundry import FoundryChatClient
from azure.identity import AzureCliCredential
from agent_framework import tool

dotenv.load_dotenv(dotenv.find_dotenv())

endpoint = os.getenv("AZURE_AI_PROJECT_ENDPOINT")
model = os.getenv("AZURE_AI_MODEL_DEPLOYMENT_NAME")

if not endpoint or not model:
    raise ValueError(
        "Missing required environment variables. "
        "Please set AZURE_AI_PROJECT_ENDPOINT and AZURE_AI_MODEL_DEPLOYMENT_NAME in your .env file."
    )

provider = FoundryChatClient(
    project_endpoint=endpoint,
    model=model,
    credential=AzureCliCredential()
)
```

## ਆਪਣਾ ਪਹਿਲਾ ਏਜੰਟ ਬਣਾਉਣਾ

ਇੱਕ ਏਜੰਟ ਨੂੰ ਦੋ ਚੀਜ਼ਾਂ ਦੀ ਲੋੜ ਹੁੰਦੀ ਹੈ:

- **ਹਦਾਇਤਾਂ** ਜੋ ਦੱਸਦੀਆਂ ਹਨ *ਉਹ ਕੌਣ ਹੈ* ਅਤੇ *ਕਿਵੇਂ ਵਰਤਣਾ ਹੈ* (ਇੱਕ ਸਿਸਟਮ ਪ੍ਰਾਂਪਟ).
- **ਟੂਲਜ਼** — ਪਾਇਥਨ ਫੰਕਸ਼ਨਾਂ ਦੇਸ਼ਾਭਿਤੀਆਂ ਜਿਹੜੀਆਂ `@tool` ਨਾਲ ਸੰਜੋਈਆਂ ਗਈਆਂ ਹਨ ਜਿਹੜਾ ਏਜੰਟ ਜਾਣਕਾਰੀ ਪ੍ਰਾਪਤ ਕਰਨ ਜਾਂ ਕਾਰਵਾਈ ਕਰਨ ਲਈ ਬੁਲਾ ਸਕਦਾ ਹੈ.

ਹੇਠਾਂ ਅਸੀਂ ਇੱਕ ਸੌਖਾ ਟੂਲ ਤਿਆਰ ਕਰਦੇ ਹਾਂ ਜੋ ਮਸ਼ਹੂਰ ਛੁੱਟੀਆਂ ਗੰਤਵਿਆਂ ਦੀ ਸੂਚੀ ਵਾਪਸ ਕਰਦਾ ਹੈ। ਜਦੋਂ ਵਰਤੋਂਕਾਰ ਯਾਤਰਾ ਦੀ ਸਿਫਾਰਿਸ਼ਾਂ ਲਈ ਪੁੱਛਦਾ ਹੈ ਤਾਂ ਐਜੰਟ ਇਹ ਟੂਲ ਵਰਤੇਗਾ।


```python
@tool(approval_mode="never_require")
def get_destinations() -> list[str]:
    """Get a list of popular vacation destinations."""
    return [
        "Barcelona",
        "Paris",
        "Berlin",
        "Tokyo",
        "Sydney",
        "New York City",
        "Cairo",
        "Cape Town",
        "Rio de Janeiro",
        "Bali",
    ]
```

```python
agent = provider.as_agent(
    name="TravelAgent",
    instructions=(
        "You are a helpful travel agent. Help users find their perfect vacation "
        "destination based on their preferences. Use the get_destinations tool "
        "to see available destinations."
    ),
    tools=[get_destinations],
)

response = await agent.run(
    "I'm looking for a warm beach destination. What do you recommend?"
)
print(response)
```

## ਸਟਰੀਮਿੰਗ ਜਵਾਬ

ਵਧੇਰੇ ਇੰਟਰਐਕਟਿਵ ਅਨੁਭਵ ਲਈ ਤੁਸੀਂ ਏਜੰਟ ਦੇ ਜਵਾਬ ਨੂੰ **ਸਟਰੀਮ** ਕਰ ਸਕਦੇ ਹੋ। ਪੂਰੇ ਜਵਾਬ ਦੇ ਇੰਤਜ਼ਾਰ ਕਰਨ ਦੀ ਬਜਾਏ, ਏਜੰਟ ਟੈਕਸਟ ਹਿੱਸੇ ਤਿਆਰ ਹੋਣ ਤੇ ਤੁਰੰਤ ਦਿੰਦਾ ਹੈ। ਇਹ ਖਾਸ ਕਰਕੇ ਚੈਟ ਇੰਟਰਫੇਸਾਂ ਵਿੱਚ ਲਾਭਦਾਇਕ ਹੁੰਦਾ ਹੈ ਜਿੱਥੇ ਤੁਸੀਂ ਰੀਅਲ ਟਾਈਮ ਵਿੱਚ ਆਊਟਪੁੱਟ ਦਿਖਾਉਣਾ ਚਾਹੁੰਦੇ ਹੋ।  


```python
async for chunk in agent.run(
    "Tell me about Tokyo as a travel destination", stream=True
):
    print(chunk, end="", flush=True)
```

## ਸੰਖੇਪ

- **ਇੱਕ ਪ੍ਰਦਾਤਾ ਬਣਾਓ** ਜੋ Microsoft Foundry Agent Service ਨਾਲ `FoundryChatClient` ਰਾਹੀਂ ਜੁੜਦਾ ਹੈ।

- **ਜਵਾਬ ਸਟ੍ਰੀਮ ਕਰੋ** ਤੁਰੰਤ ਨਤੀਜੇ ਲਈ।
