# 🔍 ಮೈಕ್ರೋಸಾಫ್ಟ್ ಫೌಂಡ್ರಿ (.NET) ಸಹಿತ ಎಂಟರ್ಪ್ರೈಸ್ RAG

## 📋 ಕಲಿಕಾ ಗುರಿಗಳು

ಈ ನೋಟ್‌ಬುಕ್ ಎಂಟರ್ಪ್ರೈಸ್-ಗ್ರೇಡ್ ರಿಟ್ರೀವಲ್-ಆಗ್ಮೆಂಟೆಡ್ ಜನರೇಷನ್ (RAG) ವ್ಯವಸ್ಥೆಗಳನ್ನು ಮೈಕ್ರೋಸಾಫ್ಟ್ ಫೌಂಡ್ರಿ ಸಹಿತ ಡಾಟ್‌ನೆಟ್‌ನಲ್ಲಿ ಮೈಕ್ರೋಸಾಫ್ಟ್ ಏಜೆಂಟ್ ಫ್ರೇಮ್‌ವರ್ಕ್ ಬಳಸಿ ಹೇಗೆ ನಿರ್ಮಿಸಬೇಕೆಂಬುದನ್ನು ತೋರಿಸುತ್ತದೆ. ನೀವು ಉತ್ಪಾದನಾ-ತಯಾರಿಯಾದ ಏಜೆಂಟ್‌ಗಳನ್ನು ರಚಿಸಲು ಕಲಿಯುತ್ತೀರಿ, ಅವು ಡಾಕ್ಯುಮೆಂಟುಗಳನ್ನು ಹುಡುಕಿ ನಿಖರ, ಪಾರિસರ-ಜಾಗೃತಿ ಉತ್ತರಗಳನ್ನು ಎಂಟರ್ಪ್ರೈಸ್ ಭದ್ರತೆ ಮತ್ತು ವ್ಯಾಪ್ತಿಯೊಂದಿಗೆ ಒದಗಿಸಬಹುದು.

**ನೀವು ನಿರ್ಮಿಸುವ ಎಂಟರ್ಪ್ರೈಸ್ RAG ಸಾಮರ್ಥ್ಯಗಳು:**
- 📚 **ಡಾಕ್ಯುಮೆಂಟ್ ಬುದ್ಧಿಮತ್ತೆ**: ಅಜೂರ್ AI ಸೇವೆಗಳೊಂದಿಗೆ ಆಧುನಿಕ ಡಾಕ್ಯುಮೆಂಟ್ ಕ್ಷೇತ್ರ ಸಂಸ್ಕರಣೆ
- 🔍 **ಸಮಾನಾರ್ಥಕ ಶೋಧ**: ಎಂಟರ್ಪ್ರೈಸ್ ವೈಶಿಷ್ಟ್ಯಗಳೊಂದಿಗೆ ಉನ್ನತ-ಕಾರ್ಯಕ್ಷಮತೆ ವೆಕ್ಟರ್ ಶೋಧ
- 🛡️ **ಭದ್ರತಾ ಒಗ್ಗಟ್ಟು**: ಪಾತ್ರ ಆಧಾರಿತ ಪ್ರವೇಶ ಮತ್ತು ಡೇಟಾ ರಕ್ಷಣೆ ಮಾದರಿಗಳು
- 🏢 **ವ್ಯಾಪಕ معماري**: ನಿಯಂತ್ರಣದೊಂದಿಗೆ ಉತ್ಪಾದನಾ-ಪ್ರತಿಸ್ತ್ರೀಯ RAG ವ್ಯವಸ್ಥೆಗಳು

## 🎯 ಎಂಟರ್ಪ್ರೈಸ್ RAG معماري

### ಮೂಲ ಎಂಟರ್ಪ್ರೈಸ್ घटಕಗಳು
- **Microsoft Foundry**: ಭದ್ರತೆ ಮತ್ತು ಅನುಕೂಲತೆಯೊಂದಿಗೆ ನಿರ್ವಹಿತ ಎಂಟರ್ಪ್ರೈಸ್ AI ವೇದಿಕೆ
- **ಸ್ಥಿರಏಜೆಂಟ್‌ಗಳು**: ಸಂಭಾಷಣಾ ಇತಿಹಾಸ ಮತ್ತು ಪಾರಿಸರ ನಿರ್ವಹಣೆಯೊಂದಿಗೆ ಸ್ಥಿತಿಗತ ಸಂಘಟಕ ಏಜೆಂಟ್‌ಗಳು
- **ವೆಕ್ಟರ್ ಸ್ಟೋರ್ ನಿರ್ವಹಣೆ**: ಎಂಟರ್ಪ್ರೈಸ್-ಗ್ರೇಡ್ ಡಾಕ್ಯುಮೆಂಟ್ ಸೂಚ್ಯಂಕ ಮತ್ತು ಪಶ್ಚಾತಪಾವತಿ
- **ಐಡೆಂಟಿಟಿ ಒಗ್ಗಟ್ಟು**: ಅಜೂರ್ AD ಪ್ರಾಮಾಣೀಕರಣ ಮತ್ತು ಪಾತ್ರ ಆಧಾರಿತ ಪ್ರವೇಶ ನಿಯಂತ್ರಣ

### .NET ಎಂಟರ್ಪ್ರೈಸ್ ಲಾಭಗಳು
- **ಟೈಪ್ ಭದ್ರತೆ**: RAG ಕಾರ್ಯಾಚರಣೆಗಳು ಮತ್ತು ಡೇಟಾ ರಚನೆಗಳಿಗೆ ಸಂಯೋಜನೆ ಸಮಯ ಪರಿಶೀಲನೆ
- **ಏಸಿಂಕ್ ಕಾರ್ಯಕ್ಷಮತೆ**: ನಿರ್ಬಂಧರಹಿತ ಡಾಕ್ಯುಮೆಂಟ್ ಸಂಸ್ಕರಣೆ ಮತ್ತು ಶೋಧ ಕಾರ್ಯಗಳು
- **ಮೆಮೊರಿ ನಿರ್ವಹಣೆ**: ದೊಡ್ಡ ಡಾಕ್ಯುಮೆಂಟ್ ಸಂಗ್ರಹಗಳಿಗೆ ಪರಿಣಾಮಕಾರಿ ಸಂಪನ್ಮೂಲ ಉಪಯೋಗ
- **ಒಗ್ಗಟ್ಟು ಮಾದರಿಗಳು**: ಆಧೃತ ಅಜೂರ್ ಸೇವಾ ಒಗ್ಗಟ್ಟು ಜೊತೆ ಅವಲಂಬನಾ ಇಂಜೆಕ್ಷನ್

## 🏗️ ತಾಂತ್ರಿಕ معماري

### ಎಂಟರ್ಪ್ರೈಸ್ RAG ಪೈಪ್ಲೈನ್
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### ಮೂಲ .NET घटಕಗಳು
- **Azure.AI.Agents.Persistent**: ಸ್ಥಿತಿ ಸ್ಥಿತಿಸ್ಥಾಪಕತೆಯೊಂದಿಗೆ ಎಂಟರ್ಪ್ರೈಸ್ ಏಜೆಂಟ್ ನಿರ್ವಹಣೆ
- **Azure.Identity**: ಭದ್ರ ಅಜೂರ್ ಸೇವಾ ಪ್ರವೇಶಕ್ಕೆ ಏಕೀಕೃತ ಪ್ರಾಮಾಣೀಕರಣ
- **Microsoft.Agents.AI.AzureAI**: ಅಜೂರ್-ಆಪ್ಟಿಮೈಜ್ಡ್ ಏಜೆಂಟ್ ಫ್ರೇಮ್‌ವರ್ಕ್ ಅನುಷ್ಟಾನ
- **System.Linq.Async**: ಉನ್ನತ-ಕಾರ್ಯಕ್ಷಮತೆಯ ಏಸಿಂಕ್ LINQ ಕಾರ್ಯಗಳು

## 🔧 ಎಂಟರ್ಪ್ರೈಸ್ ವೈಶಿಷ್ಟ್ಯಗಳು ಮತ್ತು ಲಾಭಗಳು

### ಭದ್ರತೆ ಮತ್ತು ಅನುಕೂಲತೆ
- **ಅಜೂರ್ AD ಒಗ್ಗಟ್ಟು**: ಎಂಟರ್ಪ್ರೈಸ್ ಐಡೆಂಟಿಟಿ ನಿರ್ವಹಣೆ ಮತ್ತು ಪ್ರಾಮಾಣೀಕರಣ
- **ಪಾತ್ರ ಆಧಾರಿತ ಪ್ರವೇಶ**: ಡಾಕ್ಯುಮೆಂಟ್ ಪ್ರವೇಶ ಮತ್ತು ಕಾರ್ಯಾಚರಣೆಗಳಿಗಾಗಿ ಸೂಕ್ಷ್ಮ ಅನುಮತಿಗಳು
- **ಡೇಟಾ ರಕ್ಷಣೆ**: ಸಂವೇದನಾಶೀಲ ಡಾಕ್ಯುಮೆಂಟ್‌ಗಳಿಗೆ ಆರಾಮದಲ್ಲಿನ ಮತ್ತು ಸಾರಿಗೆ ಎನ್ಕ್ರಿಪ್ಶನ್
- **ಅಡಿಟ್ ಲಾಗಿಂಗ್**: ಅನುಕೂಲತಾ ಅಗತ್ಯಗಳಿಗೆ ಸಮಗ್ರ ಚಟುವಟಿಕೆ ತರಕಾರಿ

### ಕಾರ್ಯಕ್ಷಮತೆ ಮತ್ತು ವ್ಯಾಪ್ತಿತಾಹ
- **ಕನೆಕ್ಷನ್ ಪೂಲಿಂಗ್**: ಪರಿಣಾಮಕಾರಿ ಅಜೂರ್ ಸೇವಾ ಸಂಪರ್ಕ ನಿರ್ವಹಣೆ
- **ಏಸಿಂಕ್ ಸಂಸ್ಕರಣೆ**: ಉನ್ನತ-ಪ್ರವಾಹದ ಸ್ಥಿತಿಗೆ ನಿರ್ಬಂಧರಹಿತ ಕಾರ್ಯಗಳು
- **ಕ್ಯಾಶಿಂಗ್ ತಂತ್ರಗಳು**: ಬಾರಿಗೆ ಪ್ರವೇಶಿಸಿದ ಡಾಕ್ಯುಮೆಂಟ್‌ಗಳಿಗೆ ಬುದ್ಧಿವಂತ ಕ್ಯಾಶಿಂಗ್
- **ಲೋಡ್ ಸಮತೋಲನ**: ದೊಡ್ಡ ಪ್ರಮಾಣದ ನಿಯೋಜನೆಗಳಿಗೆ ವಿತರಿತ ಸಂಸ್ಕರಣೆ

### ನಿರ್ವಹಣೆ ಮತ್ತು ಮೇಲ್ವಿಚಾರಣೆ
- **ಹೆಲ್ತ್ ಚೆಕ್ಸ್**: RAG ವ್ಯವಸ್ಥಾ ಘಟಕಗಳ ಒಳಗಿನ ಮೇಲ್ವಿಚಾರಣೆ
- **ಕಾರ್ಯಕ್ಷಮತೆ ಅಂಶಗಳು**: ಶೋಧ ಗುಣಮಟ್ಟ ಮತ್ತು ಪ್ರತಿಕ್ರಿಯೆ ಸಮಯಗಳ ಮೇಲೆ ವಿರಚನೆ
- **ದೋಷ ನಿರ್ವಹಣೆ**: ಮರು ಪ್ರಯತ್ನ ನೀತಿಗಳನ್ನು ಹೊಂದಿರುವ ಸಮಗ್ರ ಅಪವಾದ ನಿರ್ವಹಣೆ
- **ರೂಪರೇಷೆ ನಿರ್ವಹಣೆ**: ಪರಿಸರ-ನಿರ್ದಿಷ್ಟ ಸೆಟಿಂಗ್‌ಗಳ ಪರಿಶೀಲನೆಯೊಂದಿಗೆ

## 📊 ಎಂಟರ್ಪ್ರೈಸ್ RAG ಮಾದರಿಗಳು

### ಡಾಕ್ಯುಮೆಂಟ್ ನಿರ್ವಹಣಾ ಮಾದರಿಗಳು
- **ಬಲ್ಕ್ ಅಪ್‌ಲೋಡ್**: ದೊಡ್ಡ ಡಾಕ್ಯುಮೆಂಟ್ ಸಂಗ್ರಹಗಳ ಪರಿಣಾಮಕಾರಿ ಸಂಸ್ಕರಣೆ
- **ಆವರ್ತನ ಅಪ್‌ಡೇಟ್‌ಗಳು**: ನೈಜ ಸಮಯ ಡಾಕ್ಯುಮೆಂಟ್ ಸೇರ್ಪಡೆ ಮತ್ತು ಬದಲಾವಣೆ
- **ಆವೃತ್ತಿ ನಿಯಂತ್ರಣ**: ಡಾಕ್ಯುಮೆಂಟ್ ಆವೃತಿಗಳನ್ನು ಮತ್ತು ಬದಲಾವಣೆಗಳು ಹತ್ತುವು
- **ಮೆಟಾಡೇಟಾ ನಿರ್ವಹಣೆ**: ಶ್ರೀಮಂತರಾದ ಡಾಕ್ಯುಮೆಂಟ್ ಗುಣಲಕ್ಷಣಗಳು ಮತ್ತು ವರ್ಗೀಕರಣ

### ಶೋಧ ಮತ್ತು ಪಶ್ಚಾತಪಾವತಿ ಮಾದರಿಗಳು
- **ಹೈಬ್ರಿಡ್ ಶೋಧ**: ಗರಿಷ್ಠ ಫಲಿತಾಂಶಕ್ಕೆ ಸಮಾನಾರ್ಥಕ ಮತ್ತು ಕೀವರ್ಡ್ ಶೋಧ ಸಂಯೋಜನೆ
- **ಮುಖಾವರಣ ಶೋಧ**: ಬಹು-ಆಯಾಮದ ಶೋಧನ ಮತ್ತು ವರ್ಗೀಕರಣ
- **ಸಂಬಂಧದ ಶಸ್ತ್ರಚಿಕಿತ್ಸೆ**: ನಿರ್ದಿಷ್ಟ ಕ್ಷೇತ್ರಕ್ಕೆ ಹೊಂದಿದ ಕಸ್ಟಮ್ ಅಂಕಗಳು
- **ಫಲಿತಾಂಶ ವರ್ಗೀಕರಣ**: ವ್ಯಾಪಾರ ಪ್ರಾಸ್ತಾವಿಕತೆ ಒಂದುಗೂಡಿಸಿದ ಆಧುನಿಕ ವರ್ಗೀಕರಣ

### ಭದ್ರತಾ ಮಾದರಿಗಳು
- **ಡಾಕ್ಯುಮೆಂಟ್ ಮಟ್ಟದ ಭದ್ರತೆ**: ಪ್ರತಿ ಡಾಕ್ಯುಮೆಂಟ್‌ಗೆ ಸೂಕ್ಷ್ಮ ಪ್ರವೇಶ ನಿಯಂತ್ರಣ
- **ಡೇಟಾ ವರ್ಗೀಕರಣ**: ಸ್ವಯಂಚಾಲಿತ ಸಂವೇದನಾಶೀಲತೆ ಲೇಬಲಿಂಗ್ ಮತ್ತು ರಕ್ಷಣಾ
- **ಅಡಿಟ್ ಟ್ರೇಲ್‌ಗಳು**: ಎಲ್ಲಾ RAG ಕಾರ್ಯಾಚರಣೆಗಳ ಸಮಗ್ರ ಲಾಗಿಂಗ್
- **ಗೋಪ್ಯತೆ ರಕ್ಷಣೆ**: PII ಪತ್ತೆ ಮತ್ತು ಒರೆತಮನ capabilities

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-5-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
