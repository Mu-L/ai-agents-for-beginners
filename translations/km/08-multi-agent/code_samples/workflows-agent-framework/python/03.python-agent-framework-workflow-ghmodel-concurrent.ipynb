## рҹ“Ӣ бһҳбҹҒбһҡбҹҖбһ“бһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶

бһҖбҹ’бһҡбһҠбһ¶бһҹбһҖбҹҶбһҺбһҸбҹӢбһ бҹҒбһҸбһ»бһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһё **бһӮбҹҶбһҡбһјбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”** бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Microsoft Agent Frameworkбҹ” бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһҗбһ¶бһўбһ¶бһ…бһҹбһ„бҹӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҒбҹ’бһ–бһҹбҹӢ бһҸбһ¶бһҳбһҡбһҷбҹҲбһҖбһ¶бһҡбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҡбһҪбһҳбһӮбҹ’бһ“бһ¶бһҡбһ”бһҹбҹӢбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ AI бһ…бҹ’бһҡбһҫбһ“ бһҖбҹ’бһ“бһ»бһ„бһ–бҹҒбһӣбһҸбҹӮбһҳбһҪбһҷ бһҠбҹ„бһҷбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһ”бһ¶бһ“бһӣбҹ’бһ”бһҝбһ“бһҒбҹ’бһ–бһҹбҹӢ бһ“бһ·бһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһҳбһ¶бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһҪбһҡбһ”бһ„бҹ’бһҡбһ№бһ„бһ”бһ…бҹ’бһ…бҹҒбһҖбһ‘бҹҒбһҹбһҖбһҳбҹ’бһҳбһңбһёбһ’бһёбһ–бһ¶бһҺбһ·бһҮбҹ’бһҮбһҖбһҳбҹ’бһҳбһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһҠбҹ…бһҠбһ¶бһ…бҹӢбҹ”

> **бһҹбһҳбҹ’бһ—бһ¶бһҹбһ“бҹҚбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡ:** бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ“бҹҒбҹҮбһ”бһ¶бһ“бһҷбҹ„бһ„бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҳбҹүбһјбһҠбҹӮбһӣ GitHub бһҳбһ»бһ“бһ“бҹҒбҹҮбҹ” бһҳбҹүбһјбһҠбҹӮбһӣ GitHub бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһҲбһ”бҹӢ (бһ“бһ№бһ„бһ…бҹҒбһүбһ–бһёбһ”бҹ’бһҡбһҫбһ“бҹ…бһҒбҹӮбһҖбһҖбҹ’бһҖбһҠбһ¶ бҹўбҹ бҹўбҹҰ) бһҠбҹ„бһҷбһӮбҹҒбһҘбһЎбһјбһңбһ“бҹҒбҹҮбһ”бҹ’бһҡбһҫ **Microsoft Foundry** бһҸбһ¶бһҳбһҡбһҷбҹҲ `FoundryChatClient` бһҠбҹӮбһӣбһ•бҹ’бһҸбҹ„бһҸбһӣбһҫ Azure OpenAI **Responses API**бҹ”

## рҹҺҜ бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҡбҹҖбһ“

### рҹҡҖ **бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҖбһ¶бһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”**
- **бһҖбһ¶бһҡбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹ„бһҷбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”**: бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһҳбһҪбһҷбһӮбҹ’бһ“бһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҒбҹ’бһ–бһҹбҹӢбһ”бҹҶбһ•бһ»бһҸ
- **бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ”бһҺбҹ’бһҸбһ¶бһүбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡ**: бһҡбҹҖбһ”бһ…бҹҶбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҮбһ¶бһҳбһҪбһҷбһӮбҹ’бһ“бһ¶ бһ–бҹҒбһӣбһҠбҹӮбһӣбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһ—бһ¶бһ–бһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ
- **бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–**: бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһӣбҹ’бһ”бһҝбһ“бһҡбһ бҹҗбһҹбһҳбҹүбһҸбҹӢбһҸбһ¶бһҳбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҖбҹ’бһ“бһ»бһ„бһ–бҹҒбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”
- **бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ’бһ“бһ’бһ¶бһ“**: бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ’бһ“бһ’бһ¶бһ“бһҳбҹүбһјбһҢбҹӮбһӣ AI бһҠбҹ„бһҷбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһ”бһҺбҹ’бһҸбһ¶бһүбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”

### рҹҸ—пёҸ **бһӮбҹҶбһҡбһјбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”бһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢ**
- **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ”бҹҶбһ”бҹӮбһҖ-бһ…бһјбһӣбһҡбһҪбһҳ (Fork-Join)**: бһ”бҹҶбһ”бҹӮбһҖбһҖбһ¶бһҡбһ„бһ¶бһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ…бҹ’бһҡбһҫбһ“ бһ“бһ·бһ„бһ”бһүбҹ’бһ…бһјбһӣбһӣбһ‘бҹ’бһ’бһ•бһӣ
- **бһҡбһ”бҹҖбһ”бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҸбһ¶бһҳбһ”бһҺбҹ’бһҸбһ¶бһү (Pipeline Parallelism)**: бһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҠбҹӮбһӣбһӮбҹ’бһҡбһ”бҹӢбһҖбҹҶбһҡбһ·бһҸбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҡбһҹбҹӢбһҡбһңбһҫбһҖбһҮбһ¶бһ“бһ·бһ…бҹ’бһ…
- **бһҖбһ¶бһҡбһ”бҹ’бһҡбһҖбһҪбһҸбһ”бҹ’бһҡбһҮбҹӮбһ„бһ’бһ“бһ’бһ¶бһ“ (Load Balancing)**: бһ…бҹӮбһҖбһ…бһ¶бһҷбһҖбһ¶бһҡбһ„бһ¶бһҡбһІбҹ’бһҷбһҹбҹ’бһҳбһҫбһӣбҹ’бһҳбһҳбһӣбһҫбһ’бһ“бһ’бһ¶бһ“бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“
- **бһ…бҹҶбһҺбһ»бһ…бһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣ (Synchronization Points)**: бһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”бһ“бҹ…бһҖбҹҶбһҡбһ·бһҸбһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ—

### рҹҸў **бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһўбҹҒбһ“бһ’бһҫбһҡбҹүбһ¶бһҷбһҹбҹҚбһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”**
- **бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҜбһҖбһҹбһ¶бһҡбһ…бҹ’бһҡбһҫбһ“бһ‘бһ»бһҖбһҖбҹ’бһ“бһ»бһ„бһ–бҹҒбһӣбһҸбҹӮбһҳбһҪбһҷ**: бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҜбһҖбһҹбһ¶бһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһ“бһ»бһ„бһ–бҹҒбһӣбһҸбҹӮбһҳбһҪбһҷ
- **бһңбһ·бһ—бһ¶бһӮбһҒбҹ’бһӣбһ№бһҳбһҹбһ¶бһҡбһ–бҹҒбһӣбһҸбҹ’бһҡбһ„бҹӢ**: бһңбһ·бһ—бһ¶бһӮбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”бһ“бҹғбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ…бһјбһӣ
- **бһ”бһ„бҹ’бһҖбһҫбһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҮбһ¶бһҖбҹ’бһҡбһ»бһҳ**: бһӣбһҫбһҖбһҖбһҳбҹ’бһ–бһҹбҹӢбһӣбҹ’бһ”бһҝбһ“бһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһ‘бһјбһӣбҹҶбһ‘бһјбһӣбһ¶бһҷ
- **бһңбһ·бһ—бһ¶бһӮбһҳбһ»бһҒбһ„бһ¶бһҡбһ…бһҳбҹ’бһҡбһ»бҹҮ**: бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҹбһҳбҹҗбһҷбһҹбҹ’бһҡбһ”бһ“бҹғбһҳбһ¶бһҸбһ·бһҖбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ•бҹ’бһҹбҹҒбһ„бҹ— (бһўбһҸбҹ’бһҗбһ”бһ‘ бһҡбһјбһ”бһ—бһ¶бһ– бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ)

````python
import os
from typing import Any

from agent_framework import (
    Executor,
    Message,
    WorkflowBuilder,
    WorkflowContext,
    WorkflowViz,
    handler,
)
from agent_framework.foundry import FoundryChatClient
from azure.identity import AzureCliCredential
from dotenv import load_dotenv

load_dotenv()
````

````python
# Configure the Microsoft Foundry client with keyless authentication.
# FoundryChatClient targets the Azure OpenAI Responses API.
provider = FoundryChatClient(
    project_endpoint=os.environ["AZURE_AI_PROJECT_ENDPOINT"],
    model=os.environ["AZURE_AI_MODEL_DEPLOYMENT_NAME"],
    credential=AzureCliCredential(),
)
````

````python
ResearcherAgentName = "Researcher-Agent"
ResearcherAgentInstructions = "You are my travel researcher, working with me to analyze the destination, list relevant attractions, and make detailed plans for each attraction."
````

````python
PlanAgentName = "Plan-Agent"
PlanAgentInstructions = "You are my travel planner, working with me to create a detailed travel plan based on the researcher's findings."
````

````python
research_agent = provider.as_agent(
    name=ResearcherAgentName,
    instructions=ResearcherAgentInstructions,
)

plan_agent = provider.as_agent(
    name=PlanAgentName,
    instructions=PlanAgentInstructions,
)
````

````python
# A passthrough executor that broadcasts the user input to every agent in parallel.
class InputDispatcher(Executor):
    """Forward the user input unchanged to all participating agents."""

    @handler
    async def forward(self, text: str, ctx: WorkflowContext[str]) -> None:
        await ctx.send_message(text)


dispatcher = InputDispatcher(id="dispatcher")
agents = [research_agent, plan_agent]

workflow = (
    WorkflowBuilder(
        start_executor=dispatcher,
        output_executors=agents,
    )
    .add_fan_out_edges(dispatcher, agents)
    .build()
)
````

````python
print("Generating workflow visualization...")
viz = WorkflowViz(workflow)
# Print out the mermaid string.
print("Mermaid string: \n=======")
print(viz.to_mermaid())
print("=======")
# Print out the DiGraph string.
print("DiGraph string: \n=======")
print(viz.to_digraph())
print("=======")
# SVG export needs the optional graphviz extra plus the graphviz system binary;
# fall back gracefully if it is not available.
try:
    svg_file = viz.export(format="svg")
    print(f"SVG file saved to: {svg_file}")
except ImportError as e:
    svg_file = None
    print(f"SVG export skipped (install graphviz to enable): {e}")
````

````python
from IPython.display import SVG, display, HTML
import os

print(f"Attempting to display SVG file at: {svg_file}")

if svg_file and os.path.exists(svg_file):
    try:
        # Preferred: direct SVG rendering
        display(SVG(filename=svg_file))
    except Exception as e:
        print(f"вҡ пёҸ Direct SVG render failed: {e}. Falling back to raw HTML.")
        try:
            with open(svg_file, "r", encoding="utf-8") as f:
                svg_text = f.read()
            display(HTML(svg_text))
        except Exception as inner:
            print(f"вқҢ Fallback HTML render also failed: {inner}")
else:
    print("вқҢ SVG file not found. Ensure viz.export(format='svg') ran successfully.")
````

````python
events = await workflow.run("Plan a trip to Seattle in December")
outputs = events.get_outputs()
````

````python
if outputs:
    print("===== Final Aggregated Responses =====")
    # outputs is a list of AgentResponse objects, one per output executor
    # (research_agent then plan_agent), in the order given to output_executors.
    for i, response in enumerate(outputs, start=1):
        print(f"{'-' * 60}\n\n{i:02d}:\n{response.text}")
````
