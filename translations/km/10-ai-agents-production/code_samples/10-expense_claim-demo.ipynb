 ## бһҖбҹҶбһҺбһҸбҹӢбһӮбҹҶбһҡбһјбһ…бҹҶбһҺбһ¶бһҷ

 бһ”бһ„бҹ’бһҖбһҫбһҸбһӮбҹҶбһҡбһј Pydantic бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бҹҶбһҺбһ¶бһҷбһ“бһ·бһҳбһҪбһҷбҹ— бһ“бһ·бһ„бһҗбҹ’бһ“бһ¶бһҖбҹӢ ExpenseFormatter бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһҳбҹ’бһӣбҹӮбһ„бһҹбҹҶбһ“бһҪбһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ‘бҹ…бһҮбһ¶бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ…бҹҶбһҺбһ¶бһҷбһҠбҹӮбһӣбһҡбҹҖбһ”бһ…бҹҶбһҸбһ¶бһҳбһҡбһҷбҹҲбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бҹ”

 бһ…бҹҶбһҺбһ¶бһҷбһ“бһёбһҳбһҪбһҷбҹ—бһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһҸбҹҶбһҺбһ¶бһ„бһҖбҹ’бһ“бһ»бһ„бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбҹ–
 `{'date': '07-Mar-2025', 'description': 'flight to destination', 'amount': 675.99, 'category': 'Transportation'}`


```python
class Expense(BaseModel):
    date: str = Field(..., description="Date of expense in dd-MMM-yyyy format")
    description: str = Field(..., description="Expense description")
    amount: float = Field(..., description="Expense amount")
    category: str = Field(..., description="Expense category (e.g., Transportation, Meals, Accommodation, Miscellaneous)")

class ExpenseFormatter(BaseModel):
    raw_query: str = Field(..., description="Raw query input containing expense details")
    
    def parse_expenses(self) -> List[Expense]:
        """
        Parses the raw query into a list of Expense objects.
        Expected format: "date|description|amount|category" separated by semicolons.
        """
        expense_list = []
        for expense_str in self.raw_query.split(";"):
            if expense_str.strip():
                parts = expense_str.strip().split("|")
                if len(parts) == 4:
                    date, description, amount, category = parts
                    try:
                        expense = Expense(
                            date=date.strip(),
                            description=description.strip(),
                            amount=float(amount.strip()),
                            category=category.strip()
                        )
                        expense_list.append(expense)
                    except ValueError as e:
                        print(f"[LOG] Parse Error: Invalid data in '{expense_str}': {e}")
        return expense_list
```

бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһ§бһ”бһҖбһҡбһҺбҹҚбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹҠбһёбһҳбҹӮбһӣбһҠбһҫбһҳбҹ’бһ”бһёбһҠбһ¶бһҖбҹӢбһҹбҹ’бһ“бһҫбһҖбһ¶бһҡбһ‘бһ¶бһҳбһ‘бһ¶бһҡбһ…бҹҶбһҺбһ¶бһҷбҹ”
- бһ§бһ”бһҖбһҡбһҺбҹҚбһ“бҹҒбҹҮбһ”бҹ’бһҡбһҫ `@tool` бһҒбҹ’бһҹбҹӮбһҖбһ¶бһҡбһ–бһ¶бһҡбһ–бһё Microsoft Agent Frameworkбҹ”
- бһңбһ¶бһӮбһҺбһ“бһ¶бһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһҹбһҡбһ»бһ”бһ“бҹғбһ…бҹҶбһҺбһ¶бһҷбһ“бһ·бһ„бһҠбһ¶бһҖбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһҡбһҪбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҒбҹ’бһӣбһ№бһҳбһҹбһ¶бһҡбһўбҹҠбһёбһҳбҹӮбһӣбҹ”


```python
@tool(approval_mode="never_require")
def generate_expense_email(
    expense_data: Annotated[str, "Semicolon-separated expense entries in 'date|description|amount|category' format"]
) -> str:
    """Generate an email to submit an expense claim to the Finance Team."""
    formatter = ExpenseFormatter(raw_query=expense_data)
    expenses = formatter.parse_expenses()
    if not expenses:
        return "No valid expenses found to include in the email."
    total_amount = sum(e.amount for e in expenses)
    email_body = "Dear Finance Team,\n\n"
    email_body += "Please find below the details of my expense claim:\n\n"
    for e in expenses:
        email_body += f"- {e.date} | {e.description}: ${e.amount:.2f} ({e.category})\n"
    email_body += f"\nTotal Amount: ${total_amount:.2f}\n\n"
    email_body += "Receipts for all expenses are attached for your reference.\n\n"
    email_body += "Thank you,\n[Your Name]"
    return email_body
```

# бһ§бһ”бһҖбһҡбһҺбҹҚбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбһҖбһ…бҹҶбһҺбһ¶бһҷбһ’бҹ’бһңбһҫбһҠбҹҶбһҺбһҫбһҡбһ–бһёбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһ¶бһҖбҹ’бһҷбһ”бһҫбһҖбһ”бҹ’бһҡбһ¶бһҖбҹӢ

бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһ§бһ”бһҖбһҡбһҺбҹҚбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбһҖбһ…бҹҶбһҺбһ¶бһҷбһ’бҹ’бһңбһҫбһҠбҹҶбһҺбһҫбһҡбһ–бһёбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһ¶бһҖбҹ’бһҷбһ”бһҫбһҖбһ”бҹ’бһҡбһ¶бһҖбҹӢбҹ”
- бһ§бһ”бһҖбһҡбһҺбҹҚбһ“бҹҒбҹҮбһ”бҹ’бһҡбһҫ `@tool` decorator бһ–бһё Microsoft Agent Frameworkбҹ”
- бһңбһ¶бһўбһ¶бһ“бһҡбһјбһ”бһ—бһ¶бһ–бһ–бһ¶бһҖбҹ’бһҷбһ”бһҫбһҖбһ”бҹ’бһҡбһ¶бһҖбҹӢ бһҖбһјбһҠбһңбһ¶бһІбҹ’бһҷбһҮбһ¶ base64 бһ бһҫбһҷбһ•бҹ’бһҠбһӣбҹӢ URI бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҸбҹ’бһҡбһјбһ“бһҮбҹ’бһҡбһ¶бһ”бҹ”


```python
@tool(approval_mode="never_require")
def load_receipt_image(
    image_path: Annotated[str, "Path to the receipt image file"] = "receipt.jpg"
) -> str:
    """Load a receipt image and return its base64-encoded data URI for OCR extraction."""
    try:
        with open(image_path, "rb") as f:
            image_data = base64.b64encode(f.read()).decode("utf-8")
        return f"data:image/jpeg;base64,{image_data}"
    except Exception as e:
        error_msg = f"[LOG] Error loading image '{image_path}': {str(e)}"
        print(error_msg)
        return error_msg
```

## бһҖбһ¶бһҡбһ…бҹҶбһҺбһ¶бһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ

бһҖбҹҶбһҺбһҸбҹӢбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ бһ“бһ·бһ„бһ—бҹ’бһҮбһ¶бһ”бҹӢбһ–бһҪбһҖбһңбһ¶бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҹбһҖбһҳбҹ’бһҳбһ—бһ¶бһ–бһҡбҹҖбһ„бһҮбһ¶бһӣбҹҶбһҠбһ¶бһ”бҹӢбһҠбҹ„бһҷбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫ `WorkflowBuilder`бҹ”
- бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ OCR бһҠбһҖбһҹбҹ’бһҡбһ„бҹӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ…бҹҶбһҺбһ¶бһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ–бһёбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһ“бҹ’бһӣбһ№бһҖбһ”бһ„бҹӢбһ”бҹ’бһҡбһ¶бһҖбҹӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ§бһ”бһҖбһҡбһҺбҹҚ `load_receipt_image`бҹ”
- бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ бһўбҹҠбһёбһҳбҹӮбһӣ бһ‘бһ‘бһҪбһӣбһҷбһҖбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһҠбһҖбһҹбҹ’бһҡбһ„бҹӢ бһ бһҫбһҷбһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹҠбһёбһҳбҹӮбһӣбһ–бһ¶бһҖбҹ’бһҷбһ–бҹҒбһүбһ”бһ„бҹӢбһ”бҹ’бһҡбһ¶бһҖбҹӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҮбһ¶бһҮбҹҶбһ“бһ¶бһүбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ§бһ”бһҖбһҡбһҺбҹҚ `generate_expense_email`бҹ”
- `WorkflowBuilder` бһҮбһ¶бһҳбһҪбһҷбһӮбҹ’бһ“бһ¶ `add_edge` бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһүбҹ’бһҮбһёбһ”бһ“бҹ’бһҹбһӣбҹӢбһӣбҹҶбһҠбһ¶бһ”бҹӢбҹ– бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ OCR вҶ’ бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡ бһўбҹҠбһёбһҳбҹӮбһӣбҹ”


```python
ocr_agent = client.as_agent(
    tools=[load_receipt_image],
    name="OCRAgent",
    instructions=(
        "You are an expert OCR assistant specialized in extracting structured data from receipt images. "
        "Use the 'load_receipt_image' tool to load the receipt image, then analyze it and extract "
        "travel-related expense details in the format: 'date|description|amount|category' separated by semicolons. "
        "Follow these rules: "
        "- Date: Convert dates (e.g., '4/4/22') to 'dd-MMM-yyyy' (e.g., '04-Apr-2022'). "
        "- Description: Extract item names. "
        "- Amount: Use numeric values (e.g., '4.50' from '$4.50'). "
        "- Category: Infer from context (e.g., 'Meals' for food, 'Transportation' for travel, "
        "'Accommodation' for lodging, 'Miscellaneous' otherwise). "
        "Ignore totals, subtotals, or service charges unless they are itemized expenses. "
        "If no expenses are found, return 'No expenses detected'. "
        "Return only the structured data, no additional text."
    ),
)

email_agent = client.as_agent(
    name="EmailAgent",
    tools=[generate_expense_email],
    instructions=(
        "You are an expense claim email generator. Take the travel expense data from the previous agent "
        "(in 'date|description|amount|category' format separated by semicolons) and use the "
        "'generate_expense_email' tool to produce a professional expense claim email. "
        "Pass the semicolon-separated expense data directly to the tool."
    ),
)
```

## бһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһҳбҹҒ

бһҹбһ„бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҸбһ¶бһҳбһӣбҹҶбһҠбһ¶бһ”бҹӢ бһ бһҫбһҷбһҡбһҸбҹӢбһңбһ¶ бһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ„бҹ’бһҖбһ¶бһ“бҹӢбһҠбҹғ бһ“бһ·бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹҠбһёбһҳбҹӮбһӣбһўбҹҮбһўбһ¶бһ„бһ…бҹҶбһҺбһ¶бһҷбҹ”


> **бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢбҹ–** бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ’бҹ’бһңбһҫбһҖбһ¶бһҡвҖӢбһ“бҹҒбҹҮвҖӢбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“вҖӢбһ•бҹ’бһ‘бҹҒбһҡбһҡбһјбһ”бһ—бһ¶бһ–а№ғаёҡа№ҖаёӘаёЈа№ҮаёҲбһҮбһ¶вҖӢбһўбһҸбҹ’бһҗбһ”бһ‘ base64 бһҠбҹӮбһӣвҖӢбһўбҹ’бһ“бһҖбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ (бһҡбһҪбһҳбһ‘бһ¶бҹҶбһ„ gpt-4.1-mini) бһ“бһ№бһ„бһҳбһ·бһ“бһҠбһ¶бһҖбҹӢбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбһ¶бһҮбһ¶бһҸбҹ’бһҡбһјбһңбһҹбҹҶбһҡбһ¶бһ”бҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһЎбһҫбһҷбҹ”
> бһңбһ¶бһўбһ¶бһ…бһӣбһҫбһҹбһ–бһёвҖӢбһ”бһ„бҹ’бһўбһҪбһ…бһ”бһҡбһ·бһ”бһ‘бһҡбһ”бһҹбҹӢбһҳбҹүбһјбһҢбҹӮбһӣбҹ” бһҹбһјбһҳбһўбһ“бһ»бһңбһҸбҹ’бһҸ OCR бһҮбһ¶бһҳбһҪбһҷ Azure AI Vision (бһ¬бһ§бһ”бһҖбһҡбһҺбҹҚ OCR бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ) бһ бһҫбһҷбһ•бҹ’бһ‘бҹҒбһҡбһҸбҹӮбһҖбҹ’бһ“бһ»бһ„бһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ¶бһ“бһҷбһҖбһ…бҹҒбһү бһҳбһ·бһ“бһҠбһ¶бһҖбҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһ‘бҹҒбһ¬бһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣбһҠбһҫбһҳбҹ’бһ”бһёбһ•бҹ’бһүбһҫбһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶вҖӢбһҹбһ¶бһҡ `image_url`бҹ”
> бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҮбҹҖбһҹбһңбһ¶бһ„бһҖбҹҶбһ бһ»бһҹбһ”бһҡбһ·бһ”бһ‘ бһҹбһјбһҳбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҡбһјбһ”бһ—бһ¶бһ–а№ғаёҡа№ҖаёӘаёЈа№ҮаёҲбһҸбһјбһ…бһҮбһ¶бһ„ бһ¬бһҳбҹүбһјбһҢбҹӮбһӣбһҳбһ¶бһ“бһ”бһ„бҹ’бһўбһҪбһ…бһ”бһҡбһ·бһ”бһ‘бһ’бҹҶбһҮбһ¶бһ„бҹ”


```python
workflow = WorkflowBuilder(start_executor=ocr_agent) \
    .add_edge(ocr_agent, email_agent) \
    .build()

prompt = (
    "Please extract the raw text from the receipt image at 'receipt.jpg', "
    "focusing on travel expenses like dates, descriptions, amounts, and categories "
    "(e.g., Transportation, Accommodation, Meals, Miscellaneous). "
    "Then generate a professional expense claim email."
)

last_author = None
events = workflow.run(
    prompt,
    stream=True,
)
async for event in events:
    if event.type == "output" and isinstance(event.data, AgentResponseUpdate):
        update = event.data
        author = update.author_name
        if author != last_author:
            if last_author is not None:
                print()
            print(f"\n{'='*50}")
            print(f"# Agent - {author}:")
            print(f"{'='*50}")
            last_author = author
        print(update.text, end="", flush=True)
```
