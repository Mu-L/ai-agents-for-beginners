# рҹ”Қ RAG бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбһҮбһ¶бһҳбһҪбһҷ Microsoft Foundry (.NET)

## рҹ“Ӣ бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҡбҹҖбһ“

бһҹбҹҖбһңбһ—бҹ…бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһүбһ–бһёбһҡбһ”бҹҖбһ”бһҹбһ¶бһ„бһҹбһ„бҹӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ•бһӣбһ·бһҸбһҖбһҳбҹ’бһҳбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҠбҹ„бһҷбһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһҹбҹ’бһңбҹӮбһ„бһҡбһҖ (RAG) бһ“бҹ…бһҖбһҳбҹ’бһҡбһ·бһҸбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҹбҹҠбһ»бһҳбһ•бҹ’бһ“бҹӮбһҖбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҡбһ”бһҹбҹӢ Microsoft бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ .NET бһҮбһ¶бһҳбһҪбһҷ Microsoft Foundryбҹ” бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ»бһҒбһҡбһ”бһҡ бһҠбҹӮбһӣбһўбһ¶бһ…бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҜбһҖбһҹбһ¶бһҡбһ“бһ¶бһ“бһ¶ бһ бһҫбһҷбһ•бҹ’бһҸбһӣбҹӢбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ бһ“бһ·бһ„бһҳбһ¶бһ“бһ”бһҡбһ·бһ”бһ‘ бһҮбһ¶бһҳбһҪбһҷбһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ– бһ“бһ·бһ„бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ“бҹғбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбҹ”

**бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ– RAG бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбҹ–**
- рҹ“ҡ **бһ”бһүбҹ’бһүбһ¶бһҜбһҖбһҹбһ¶бһҡ**: бһҖбһ¶бһҡбһ”бҹ’бһҡбһҳбһјбһӣ бһ“бһ·бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҜбһҖбһҹбһ¶бһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҹбҹҒбһңбһ¶ Azure AI бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“
- рҹ”Қ **бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҳбһ¶бһ“бһўбһҸбҹ’бһҗбһ“бҹҗбһҷ**: бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбҹүбһ·бһ…бһ‘бҹҗбһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҒбҹ’бһ–бһҹбҹӢбһ”бҹҶбһ–бһ¶бһҖбҹӢбһҠбҹ„бһҷбһӣбһҖбҹ’бһҒбһҺбҹҲбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹ
- рҹӣЎпёҸ **бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–**: бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ…бһјбһӣбһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҸбһҪбһ“бһ¶бһ‘бһё бһ“бһ·бһ„бһӮбҹҶбһҡбһјбһҖбһ¶бһҡбһҖбһ¶бһҡбһ–бһ¶бһҡвҖӢбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ
- рҹҸў **бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳбһўбһ¶бһ…бһ–бһ„бҹ’бһҡбһёбһҖбһ”бһ¶бһ“**: бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’ RAG бһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһ•бһӣбһ·бһҸбһҖбһҳбҹ’бһҳ бһ“бһ·бһ„бһҳбһ¶бһ“бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷ

## рҹҺҜ бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳ RAG бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹ

### бһҹбһҳбһ¶бһҹбһ’бһ¶бһҸбһ»бһӣбҹҒбһҒбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹ
- **Microsoft Foundry**: бһңбҹҒбһ‘бһ·бһҖбһ¶ AI бһҡбһ”бһҹбҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбһҠбҹӮбһӣбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҠбҹ„бһҷбһҳбһ¶бһ“бһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбһ¶бһҡбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһ”бһ‘бһ”бһүбҹ’бһҮбһ¶
- **бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһ–бһ№бһ„бһ•бҹ’бһўбҹӮбһҖ**: бһ—бҹ’бһ“бһ¶бһҖбҹӢбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһҡбһҠбҹ’бһӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶ бһ“бһ·бһ„бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ”бһҡбһ·бһ”бһ‘
- **бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ•бҹ’бһ‘бһ»бһҖбһңбҹүбһ·бһ…бһ‘бҹҗбһҡ**: бһҖбһ¶бһҡбһҸбҹҶбһҺбҹҶбһ бһҡбһҜбһҖбһҹбһ¶бһҡбһҡбһ”бһҹбҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹбһӣбҹ’бһўбһ”бҹ’бһҡбһҹбһҫбһҡ бһ“бһ·бһ„бһҖбһ¶бһҡбһҹбҹ’бһңбҹӮбһ„бһҡбһҖ
- **бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҸбһҹбһүбҹ’бһүбһ¶бһҺ**: бһҖбһ¶бһҡбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢ Azure AD бһ“бһ·бһ„бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ…бһјбһӣбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҸбһҪбһ“бһ¶бһ‘бһё

### бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚ .NET бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһӮбҹ’бһҡбһ¶бһҹ
- **бһҹбһ»бһңбһҸбҹ’бһҗбһ·бһ—бһ¶бһ–бһ”бҹ’бһҡбһ—бҹҒбһ‘**: бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ–бҹҒбһӣбһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҡбһ”бһҹбҹӢ RAG бһ“бһ·бһ„бһҹбҹҶбһҺбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ
- **бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһўбҹҒбһҹбһёбһ„**: бһҖбһ¶бһҡбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҜбһҖбһҹбһ¶бһҡбһ“бһ·бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҠбҹӮбһӣбһҳбһ·бһ“бһ”бһ·бһ‘бһҒбҹ’бһҹбҹӮбһҖбһ¶бһҡбһ„бһ¶бһҡ
- **бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ…бһ„бһ…бһ¶бҹҶ**: бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ’бһ“бһ’бһ¶бһ“бһҷбҹүбһ¶бһ„бһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҜбһҖбһҹбһ¶бһҡбһ’бҹҶбһ“бһ¶бһ“бһ¶
- **бһӮбҹҶбһҡбһјбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣ**: бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҹбҹҒбһңбһ¶ Azure бһҠбҹ„бһҷбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ…бһ¶бһҖбҹӢбһҹбҹ„бһҠбһ¶бһҖбҹӢбһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-5-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
