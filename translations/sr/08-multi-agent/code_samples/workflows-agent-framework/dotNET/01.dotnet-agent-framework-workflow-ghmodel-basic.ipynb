# 🔄 Основни токови рада агената са Azure OpenAI (Responses API) (.NET)

## 📋 Туторијал за оркестрацију тока рада

Овај нотебук демонстрира како изградити софистициране **токове рада агената** користећи Microsoft Agent Framework за .NET и Azure OpenAI (Responses API). Научићете како да креирате пословне процесе са више корака у којима AI агенти сарађују да би обавили сложене задатке кроз структуиране обрасце оркестрације.

## 🎯 Циљеви учења

### 🏗️ **Основе архитектуре тока рада**
- **Workflow Builder**: Дизајнирање и оркестрација сложених AI процеса са више корака
- **Координација агената**: Координација више специјализованих агената у токовима рада
- **Azure OpenAI (Responses API)**: Коришћење Azure OpenAI Responses API у токовима рада
- **Визуелни дизајн тока рада**: Креирање и визуализација структура тока рада ради бољег разумевања

### 🔄 **Обрасци оркестрације процеса**
- **Секвенцијална обрада**: Повезивање више задатака агената у логичком редоследу
- **Управљање стањем**: Одржавање контекста и протока података кроз фазе тока рада
- **Обрада грешака**: Имплементација робусног опоравка од грешака и отпорности тока рада
- **Оптимизација перформанси**: Дизајн ефикасних токова рада за рад у стамбеним размерама

### 🏢 **Примене токова рада у предузећима**
- **Аутоматизација пословних процеса**: Аутоматизовање сложених организационих токова рада
- **Производни ток садржаја**: Уреднички токови рада са прегледом и фазама одобравања
- **Аутоматизација корисничке службе**: Решење захтева купаца у више корака
- **Токови обраде података**: ETL токови рада са AI-ом покретаном трансформацијом

### 🔑 **Конфигурација Azure OpenAI**

**Подешавање окружења (.env фајл):**
```env
AZURE_OPENAI_ENDPOINT=https://<your-resource>.openai.azure.com
AZURE_OPENAI_DEPLOYMENT=gpt-4.1-mini
```

**Приступ Azure OpenAI:**
1. Креирајте Azure OpenAI ресурс у Azure порталу
2. Деплојујте модел (нпр. `gpt-4.1-mini`) и запамтите име имплементације
3. Пријавите се са `az login` и конфигуришите променљиве окружења као што је приказано изнад

### 🏗️ **Преглед архитектуре тока рада**

```mermaid
graph TD
    A[Грађевински алат за радни ток] --> B[Регистар агената]
    B --> C[Покретач извршења радног тока]
    C --> D[Agent 1: Генератор садржаја
    C --> E[Agent 2: Рецензент садржаја 
    D --> F[Резултати радног тока]
    E --> F
    G[Azure OpenAI (Responses API)] --> D
    G --> E
```

**Кључне компоненте:**
- **WorkflowBuilder**: Главни мотор за оркестрацију и дизајн токова рада
- **AIAgent**: Појединачни специјализовани агенти са специфичним способностима
- **Azure OpenAI Client**: Интеграција са Azure OpenAI Responses API-јем
- **Execution Context**: Управљање стањем и протоком података између фаза тока рада

````polyglot-notebook
using System;
using System.ComponentModel;
using Azure.AI.OpenAI;
using Azure.Identity;
using Microsoft.Extensions.AI;
using Microsoft.Agents.AI;
using Microsoft.Agents.AI.Workflows;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
// Azure OpenAI with the Responses API (stable v1 endpoint). Sign in with `az login`.
var azureEndpoint = Environment.GetEnvironmentVariable("AZURE_OPENAI_ENDPOINT") ?? throw new InvalidOperationException("AZURE_OPENAI_ENDPOINT is not set.");
var deployment = Environment.GetEnvironmentVariable("AZURE_OPENAI_DEPLOYMENT") ?? "gpt-4.1-mini";
````

````polyglot-notebook
// The Azure OpenAI client is created directly from the endpoint and Azure CLI credential — no custom client options are required.
````

````polyglot-notebook
var azureClient = new AzureOpenAIClient(new Uri(azureEndpoint), new AzureCliCredential());
````

````polyglot-notebook
const string ReviewerAgentName = "Concierge";
const string ReviewerAgentInstructions = @"
    You are an are hotel concierge who has opinions about providing the most local and authentic experiences for travelers.
    The goal is to determine if the front desk travel agent has recommended the best non-touristy experience for a traveler.
    If so, state that it is approved.
    If not, provide insight on how to refine the recommendation without using a specific example. ";
````

````polyglot-notebook
const string FrontDeskAgentName = "FrontDesk";
const string FrontDeskAgentInstructions = @"""
    You are a Front Desk Travel Agent with ten years of experience and are known for brevity as you deal with many customers.
    The goal is to provide the best activities and locations for a traveler to visit.
    Only provide a single recommendation per response.
    You're laser focused on the goal at hand.
    Don't waste time with chit chat.
    Consider suggestions when refining an idea.
    """;
````

````polyglot-notebook
AIAgent reviewerAgent = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:ReviewerAgentName,instructions:ReviewerAgentInstructions);
AIAgent frontDeskAgent  = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:FrontDeskAgentName,instructions:FrontDeskAgentInstructions);
````

````polyglot-notebook
var workflow = new WorkflowBuilder(frontDeskAgent)
            .AddEdge(frontDeskAgent, reviewerAgent)
            .Build();
````

````polyglot-notebook
ChatMessage userMessage = new ChatMessage(ChatRole.User, [
	new TextContent("I would like to go to Paris.") 
]);
````

````polyglot-notebook
StreamingRun run = await InProcessExecution.StreamAsync(workflow, userMessage);
````

````polyglot-notebook
await run.TrySendMessageAsync(new TurnToken(emitEvents: true));
string id="";
string messageData="";
await foreach (WorkflowEvent evt in run.WatchStreamAsync().ConfigureAwait(false))
{
    if (evt is AgentRunUpdateEvent executorComplete)
    {
        if(id=="")
        {
            id=executorComplete.ExecutorId;
        }
        if(id==executorComplete.ExecutorId)
        {
            messageData+=executorComplete.Data.ToString();
        }
        else
        {
            id=executorComplete.ExecutorId;
        }
        // Console.WriteLine($"{executorComplete.ExecutorId}: {executorComplete.Data}");
    }
}

Console.WriteLine(messageData);
````
