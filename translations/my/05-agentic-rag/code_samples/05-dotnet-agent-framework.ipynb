# 🔍 Microsoft Foundry (.NET) နှင့်အတူ စီးပွားရေးအဆင့် RAG

## 📋 သင်ယူရန်ရည်ရွယ်ချက်များ

ဤနိုက်ဘုတ်ဘွတ်မှာ Microsoft Foundry တွင် .NET ၏ Microsoft Agent Framework ကို အသုံးပြု၍ စီးပွားရေးအဆင့် Retrieval-Augmented Generation (RAG) စနစ်များကို ဘယ်လိုတည်ဆောက်ရမည်ကို ပြသပါသည်။ သင်သည် စာရွက်စာတမ်းများကို ရှာဖွေကာ တိကျမှန်ကန်ပြီး အကြောင်းအရာကို နားလည်သည့် အဖြေများကို စီးပွားရေး လုံခြုံရေးနှင့် တိုးချဲ့နိုင်မှုဖြင့် ထုတ်ပေးနိုင်သော ထုတ်လုပ်နိုင်သော အေးဂျင့်များ ဖန်တီးနည်းကို သင်ယူမည်ဖြစ်သည်။

**သင် တည်ဆောက်မည့် စီးပွားရေး RAG စွမ်းဆောင်ရည်များ -**
- 📚 **စာရွက်စာတမ်း သိပ္ပံပညာ**: Azure AI ဝန်ဆောင်မှုများဖြင့် နည်းပြွန်စာရွက်စာတမ်း ကူးစက်မှု
- 🔍 **ဓာတ်ပြပှတ်ရှာဖွေရေး**: စီးပွားရေး လက္ခဏာများပါရှိသည့် မြန်နှုန်းမြင့် ဗက်တာရှာရန်
- 🛡️ **လုံခြုံရေးပေါင်းစပ်မှု**: နာမည်အလိုက် ဝင်ရောက်ခွင့်နှင့် ဒေတာကာကွယ်ရေး ပုံစံများ
- 🏢 **တိုးချဲ့နိုင်သော အင်ဂျင်နီယာ**: ထုတ်လုပ်နိုင်သော RAG စနစ်များနှင့် မော်နီတာလုပ်ခြင်း

## 🎯 စီးပွားရေး RAG ပုံမှန်သတ်မှတ်ချက်

### အဓိက စီးပွားရေးအစိတ်အပိုင်းများ
- **Microsoft Foundry**: လုံခြုံရေးနှင့်လိုက်နာမှုပါသော စီမံခန့်ခွဲထားသော စီးပွားရေး AI ပလက်ဖောင်း
- **မရပ်တန့်သော အေးဂျင့်များ**: စကားပြောမှတ်တမ်းနှင့် အကြောင်းအရာ စီမံခန့်ခွဲမှု ပါရှိသော အခြေအနေရှိသော အေးဂျင့်များ
- **ဗက်တာ စတိုး မန်နေဂျရီမန့်**: စီးပွားရေးအဆင့် စာရွက်စာတမ်း အညွှန်းခြင်းနှင့် ရယူမှု
- **အတည်ပြု မိတ်ဖက်ခြင်း**: Azure AD အတည်ပြုခြင်းနှင့် နာမည်အလိုက် ဝင်ရောက်ခွင့် ထိန်းချုပ်မှု

### .NET စီးပွားရေး အကျိုးအာနိသင်များ
- **အမျိုးအစား လုံခြုံမှု**: RAG လုပ်ငန်းများနှင့် ဒေတာဖွဲ့စည်းမှုများအတွက် ကိုင်တွယ်စစ်ဆေးခြင်း အချိန်ကောင်း
- **အဆက်မပြတ်စွမ်းဆောင်မှု**: စာရွက်စာတမ်း လုပ်ငန်းစဉ်များနှင့် ရှာဖွေရေး လုပ်ငန်းများတားဆီးမှုမရှိမှု
- **မှတ်ဉာဏ် စီမံခန့်ခွဲမှု**: များပြားသော စာရွက်စာတမ်းစုဆောင်းမှုများအတွက် ထိရောက်စွာ ကုန်ကျစရိတ်သုံးမှု
- **ပေါင်းစည်းမှု ပုံစံများ**: အရင်းအမြစ်ထည့်သွင်းမှုဖြင့် ဒေသခံ Azure ဝန်ဆောင်မှု ပေါင်းစည်းမှု

## 🏗️ နည်းပညာ ပုံမှန်သတ်မှတ်ချက်

### စီးပွားရေး RAG ပိုင်းလိုင်း
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### အဓိက .NET အစိတ်အပိုင်းများ
- **Azure.AI.Agents.Persistent**: အခြေအနေ ထောက်ခံမှုနှင့် စီးပွားရေးအေးဂျင့်စီမံခန့်ခွဲမှု
- **Azure.Identity**: လုံခြုံသော Azure ဝန်ဆောင်မှု ဝင်ရောက်မှုအတွက် တစ်ခုတည်းဖြင့် အတည်ပြုခြင်း
- **Microsoft.Agents.AI.AzureAI**: Azure အတွက် အဆင်ပြေသော အေးဂျင့် ဖရိမ်ဝန်း အကောင်အထည်ဖော်မှု
- **System.Linq.Async**: မြန်ဆန်သော အဆက်မပြတ် LINQ လုပ်ဆောင်မှုများ

## 📊 စီးပွားရေး RAG နမူနာများ

### စာရွက်စာတမ်း စီမံခန့်ခွဲမှု နမူနာများ
- **အုပ်စု တင်သွင်းခြင်း**: ကြီးမားသော စာရွက်စာတမ်းစုဆောင်းမှုများ ထိရောက်စွာ ပြုလုပ်ခြင်း
- **တိုးချဲ့ထားသည့် အပ်ဒိတ်များ**: အချိန်နှင့်တပြေးညီ စာရွက်စာတမ်း ထည့်သွင်းခြင်းနှင့် ပြင်ဆင်ခြင်း
- **ဗားရှင်း ထိန်းချုပ်မှု**: စာရွက်စာတမ်း ဗားရှင်းများနှင့် ပြောင်းလဲမှု အကောင်အထည်ဖော်ခြင်း
- ** မှတ်တမ်း တမ်းတမ်း စီမံခန့်ခွဲမှု**: စာရွက်စာတမ်း အမျိုးမျိုးသော အချက်အလက်များနှင့် နမူနာဇယားများ

### ရှာဖွေမှုနှင့် ရယူမှု နမူနာများ
- **ဟိုက်ဘရစ် ရှာဖွေမှု**: အဓိပ္ပါယ်နှင့် စကားလုံး ရှာဖွေရေး ပေါင်းစပ်ခြင်းဖြင့် အကောင်းဆုံးရလဒ်
- **မျက်နှာဖုံး ရှာဖွေမှု**: အလွှာများစုံဖြင့် စစ်ထုတ်ခြင်းနှင့် အမျိုးအစားခွဲခြားခြင်း
- **သင့်လျော်မှု အတိအကျ ဆက်တင်**: ဒိုမိန်းအလိုက် ပြုလုပ်သည့် အမှတ်ပေး နည်းပညာများ
- **ရလဒ် အဆင့်သတ်မှတ်ခြင်း**: စီးပွားရေး စည်းမျဉ်းများ ပေါင်းစပ်ပြီး တိုးတက်သော အဆင့်ချထားမှု

### လုံခြုံရေး ပုံစံများ
- **စာရွက်စာတမ်းအဆင့် လုံခြုံရေး**: စာရွက်စာတမ်းအလိုက် အသေးစိတ် ဝင်ရောက်ခွင့် ထိန်းချုပ်မှု
- **ဒေတာ စီမံခန့်ခွဲမှု**: ကိုယ်ရေးအချက်အလက် အာရုံစိုက်ခြင်းနှင့် ကာကွယ်မှု တိုက်ရိုက်ကာကွယ်မှု
- **စစ်ဆေး မှတ်တမ်းများ**: RAG လုပ်ငန်းတွင်း လုပ်ငန်းစဉ်များအားလုံးအတွက် နက်ရှိုင်းစွာ မှတ်တမ်းတင်ခြင်း

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-5-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
