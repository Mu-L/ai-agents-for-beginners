# 🔄 Mizunguko ya Kazi ya Wakala wa Msingi na Azure OpenAI (API za Majibu) (.NET)

## 📋 Mafunzo ya Kuendesha Mizunguko ya Kazi

Daftari hili linaonyesha jinsi ya kujenga **mizunguko ya kazi ya wakala** yenye ustadi kwa kutumia Mfumo wa Wakala wa Microsoft kwa .NET na Azure OpenAI (API za Majibu). Utajifunza kuunda mchakato wa hatua nyingi wa biashara ambapo mawakala wa AI hushirikiana kutekeleza majukumu magumu kupitia mifumo ya kuandaa mizunguko iliyopangwa vizuri.

## 🎯 Malengo ya Kujifunza

### 🏗️ **Msingi wa Usanifu wa Mizunguko ya Kazi**
- **Mtengenezaji wa Mizunguko**: Tengeneza na panga mizunguko tata ya hatua nyingi za AI
- **Uratibu wa Wakala**: Ratibu mawakala maalum wengi ndani ya mizunguko ya kazi
- **Azure OpenAI (API za Majibu)**: Tumia Azure OpenAI Responses API katika mizunguko ya kazi
- **Ubunifu wa Mizunguko kwa Mtazamo**: Unda na uonyeshe miundo ya mizunguko kwa ufahamu bora

### 🔄 **Mifumo ya Kuandaa Mizunguko ya Kazi**
- **Uchakavu wa Kusindika**: Unganisha majukumu ya wakala mengi kwa mpangilio wa kimantiki
- **Usimamizi wa Hali**: Dumisha muktadha na mtiririko wa data kati ya hatua za mizunguko
- **Ushughulikiaji wa Makosa**: Tekeleza urejeshaji wa makosa kwa uimara wa mizunguko
- **Uboreshaji wa Utendaji**: Tengeneza mizunguko yenye ufanisi kwa shughuli za viwanda vikubwa

### 🏢 **Maombi ya Mizunguko ya Biashara**
- **Uendeshaji wa Mchakato wa Biashara**: Endesha mizunguko tata ya shirika otomatiki
- **Mtiririko wa Uzalishaji wa Maudhui**: Mizunguko ya uhariri na hatua za ukaguzi na idhini
- **Uendeshaji wa Huduma kwa Wateja**: Utatuzi wa maswali ya wateja kwa hatua nyingi
- **Mizunguko ya Usindikaji Data**: Mizunguko ya ETL yenye mabadilishano yaliyoendeshwa na AI

## ⚙️ Mahitaji ya Msingi & Usanidi

```xml
<!-- Configuration and Environment -->
<PackageReference Include="DotNetEnv" Version="3.1.1" />
```

### 🔑 **Usanidi wa Azure OpenAI**

**Usanidi wa Mazingira (faili la .env):**
```env
AZURE_OPENAI_ENDPOINT=https://<your-resource>.openai.azure.com
AZURE_OPENAI_DEPLOYMENT=gpt-5-mini
```

**Ufikiaji wa Azure OpenAI:**
1. Unda rasilimali ya Azure OpenAI katika jukwaa la Azure
2. Weka mfano (kwa mfano, `gpt-5-mini`) na kumbuka jina la usambazaji
3. Ingia kwa kutumia `az login` na sanidi vigezo vya mazingira kama vilivyo hapo juu

### 🏗️ **Muhtasari wa Usanifu wa Mizunguko ya Kazi**

```mermaid
graph TD
    A[Mjenzi wa Mtiririko wa Kazi] --> B[Usajili wa Wakala]
    B --> C[Injini ya Utekelezaji wa Mtiririko wa Kazi]
    C --> D[Agent 1: Mtoaji wa Maudhui]
    C --> E[Agent 2: Mkaguzi wa Maudhui] 
    D --> F[Matokeo ya Mtiririko wa Kazi]
    E --> F
    G[Azure OpenAI (API ya Majibu)] --> D
    G --> E
```

**Vipengele Muhimu:**
- **WorkflowBuilder**: Injini kuu ya kuandaa mizunguko ya kazi
- **AIAgent**: Wakala binafsi maalum wenye uwezo fulani
- **Azure OpenAI Client**: Uunganisho wa Azure OpenAI Responses API
- **Execution Context**: Husimamia hali na mtiririko wa data kati ya hatua za mizunguko

````polyglot-notebook
using System;
using System.ComponentModel;
using Azure.AI.OpenAI;
using Azure.Identity;
using Microsoft.Extensions.AI;
using Microsoft.Agents.AI;
using Microsoft.Agents.AI.Workflows;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
// Azure OpenAI with the Responses API (stable v1 endpoint). Sign in with `az login`.
var azureEndpoint = Environment.GetEnvironmentVariable("AZURE_OPENAI_ENDPOINT") ?? throw new InvalidOperationException("AZURE_OPENAI_ENDPOINT is not set.");
var deployment = Environment.GetEnvironmentVariable("AZURE_OPENAI_DEPLOYMENT") ?? "gpt-5-mini";
````

````polyglot-notebook
// The Azure OpenAI client is created directly from the endpoint and Azure CLI credential — no custom client options are required.
````

````polyglot-notebook
var azureClient = new AzureOpenAIClient(new Uri(azureEndpoint), new AzureCliCredential());
````

````polyglot-notebook
const string ReviewerAgentName = "Concierge";
const string ReviewerAgentInstructions = @"
    You are an are hotel concierge who has opinions about providing the most local and authentic experiences for travelers.
    The goal is to determine if the front desk travel agent has recommended the best non-touristy experience for a traveler.
    If so, state that it is approved.
    If not, provide insight on how to refine the recommendation without using a specific example. ";
````

````polyglot-notebook
const string FrontDeskAgentName = "FrontDesk";
const string FrontDeskAgentInstructions = @"""
    You are a Front Desk Travel Agent with ten years of experience and are known for brevity as you deal with many customers.
    The goal is to provide the best activities and locations for a traveler to visit.
    Only provide a single recommendation per response.
    You're laser focused on the goal at hand.
    Don't waste time with chit chat.
    Consider suggestions when refining an idea.
    """;
````

````polyglot-notebook
AIAgent reviewerAgent = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:ReviewerAgentName,instructions:ReviewerAgentInstructions);
AIAgent frontDeskAgent  = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:FrontDeskAgentName,instructions:FrontDeskAgentInstructions);
````

````polyglot-notebook
var workflow = new WorkflowBuilder(frontDeskAgent)
            .AddEdge(frontDeskAgent, reviewerAgent)
            .Build();
````

````polyglot-notebook
ChatMessage userMessage = new ChatMessage(ChatRole.User, [
	new TextContent("I would like to go to Paris.") 
]);
````

````polyglot-notebook
StreamingRun run = await InProcessExecution.StreamAsync(workflow, userMessage);
````

````polyglot-notebook
await run.TrySendMessageAsync(new TurnToken(emitEvents: true));
string id="";
string messageData="";
await foreach (WorkflowEvent evt in run.WatchStreamAsync().ConfigureAwait(false))
{
    if (evt is AgentRunUpdateEvent executorComplete)
    {
        if(id=="")
        {
            id=executorComplete.ExecutorId;
        }
        if(id==executorComplete.ExecutorId)
        {
            messageData+=executorComplete.Data.ToString();
        }
        else
        {
            id=executorComplete.ExecutorId;
        }
        // Console.WriteLine($"{executorComplete.ExecutorId}: {executorComplete.Data}");
    }
}

Console.WriteLine(messageData);
````
