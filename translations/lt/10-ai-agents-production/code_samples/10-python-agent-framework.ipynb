- **Stebimąją metriką** — laikų ir registravimo pridėjimą prie agentų sąveikų
- **Vertinimą** — naudojant vertintojų agentą atsakymų kokybės įvertinimui
- **Sąnaudų valdymą** — strategijas, kaip optimizuoti atmintukus ir pasirinkti modelį

Scenarijus – **kelionių agentas**, padedantis planuoti keliones, su stebėjimu ir vertinimu viršuje.

```python
import logging
logging.getLogger("agent_framework.foundry").setLevel(logging.ERROR)

import os
import asyncio
import time
import dotenv
from typing import Annotated

from agent_framework import tool
from agent_framework.foundry import FoundryChatClient
from azure.identity import DefaultAzureCredential

dotenv.load_dotenv()

endpoint = os.getenv("AZURE_AI_PROJECT_ENDPOINT")
deployment_name = os.getenv("AZURE_AI_MODEL_DEPLOYMENT_NAME")

missing = [k for k, v in {
    "AZURE_AI_PROJECT_ENDPOINT": endpoint,
    "AZURE_AI_MODEL_DEPLOYMENT_NAME": deployment_name
}.items() if not v]

if missing:
    raise ValueError(
        f"Missing required environment variables: {', '.join(missing)}. "
        "Please set them as environment variables (e.g., in your .env file or shell environment)."
    )
```

```python
# Create the Microsoft Foundry client
client = FoundryChatClient(
    project_endpoint=endpoint,
    model=deployment_name,
    credential=DefaultAzureCredential()
)
```

## Gamybos svarstymai

Perkėlimas AI agentų iš prototipų į gamybą reikalauja atidumo trims stulpams:

1. **Stebėtinumas** — Jums reikia matomumo, ką agentas daro, kiek laiko tai užtrunka ir kokius įrankius jis naudoja. Be sekimo ir registravimo, trikčių šalinimas gamyboje beveik neįmanomas.

2. **Vertinimas** — Automatizuoti kokybės patikrinimai užtikrina, kad agento atsakymai išliktų tikslūs, išsamūs ir naudingi laikui bėgant. Įvertinimo agentas gali įvertinti atsakymus pagal apibrėžtus kriterijus.

3. **Išlaidų valdymas** — Žetonų naudojimas tiesiogiai veikia išlaidas. Strategijos, tokios kaip užklausų optimizavimas, modelio parinkimas ir talpinimas, padeda kontroliuoti išlaidas nekeliant kokybės kompromiso.


## Stebimo agento kūrimas

Apibrėžiame kelionių įrankius ir apgaubiam agento kvietimą laiko matavimu, kad galėtume stebėti delsą. Gamyboje integruotumėte su OpenTelemetry arba panašia sekimo sistema.


```python
@tool(approval_mode="never_require")
def get_flight_info(destination: Annotated[str, "The destination city"]) -> str:
    """Get flight information for a destination."""
    flights = {
        "Paris": "BA 304, 08:30-11:45, $350",
        "Tokyo": "JL 044, 11:00-07:00+1, $890",
        "Barcelona": "VY 7821, 07:15-10:30, $280",
    }
    return flights.get(destination, f"No flights found to {destination}")


@tool(approval_mode="never_require")
def get_activity_suggestions(destination: Annotated[str, "The destination city"]) -> str:
    """Get activity suggestions for a destination."""
    activities = {
        "Paris": "Louvre Museum, Eiffel Tower, Seine River Cruise, Montmartre walking tour",
        "Tokyo": "Senso-ji Temple, Tsukiji Market tour, Shibuya Crossing, teamLab Borderless",
        "Barcelona": "Sagrada Familia, Park Güell, La Boqueria Market, Gothic Quarter walk",
    }
    return activities.get(destination, f"No activities found for {destination}")
```

```python
agent = client.as_agent(
    tools=[get_flight_info, get_activity_suggestions],
    name="TravelAgent",
    instructions="You are a helpful travel agent. Use the available tools to help users plan their trips. Provide comprehensive, actionable travel advice.",
)

# Simple observability: track timing
start_time = time.time()
response = await agent.run(
    "I want to plan a day trip in Paris. What flights and activities do you recommend?",
    )
elapsed = time.time() - start_time
print(f"Response ({elapsed:.2f}s):\n{response}")
```

## Vertinimo modeliai

Bendras gamybos modelis yra naudoti antrą agentą kaip **vertintoją**. Vertintojas įvertina pagrindinio agente atsakymą pagal iš anksto nustatytus kriterijus, tokius kaip išsamumas, tikslumas ir naudingumas.

Tai leidžia:
- Automatinius kokybės patikrinimus, kol atsakymai pasiekia naudotojus
- Atpažinti regresiją, kai keičiasi užklausos ar modeliai
- Nuolatinį agento našumo stebėjimą laikui bėgant


```python
evaluator = client.as_agent(
    name="ResponseEvaluator",
    instructions="""You evaluate travel agent responses on these criteria:
1. Completeness (1-5): Did it cover flights AND activities?
2. Accuracy (1-5): Is the information consistent?
3. Helpfulness (1-5): Would a traveler find this actionable?
4. Overall Score (1-5)
Provide scores and a brief explanation for each.""",
)

evaluation = await evaluator.run(f"Evaluate this travel agent response:\n\n{response}")
print(f"Evaluation:\n{evaluation}")
```

## Kaštų valdymo strategijos

Kaštų kontrolė yra labai svarbi gamybos AI agentams. Štai pagrindinės strategijos:

| Strategija | Aprašymas |
|---|---|
| **Užuominos optimizavimas** | Išlaikykite sistemos instrukcijas glaustas. Pašalinkite perteklinį kontekstą, kad sumažintumėte įvesties žetonų skaičių. |
    "| **Modelio pasirinkimas** | Paprastoms užduotims, tokioms kaip klasifikavimas ar išgavimas, naudokite mažesnius, pigesnius modelius (pvz., GPT-5-mini), o sudėtingoms užduotims rezervuokite didesnius modelius. |\n",
| **Talpinimas (Caching)** | Talpinkite įrankių rezultatus ir dažnus užklausimus, kad išvengtumėte pasikartojančių API kvietimų. |
| **Žetonų biudžetai** | Nustatykite `max_tokens` ribas, kad būtų išvengta netikėtai ilgų atsakymų. |
| **Dėžių grupavimas (Batching)** | Jei įmanoma, sugrupuokite kelias vartotojo užklausas į vieną API kvietimą. |

Praktikoje gerai veikia daugiasluoksnis metodas: paprastas užklausas nukreipkite greitam, nebrangiam modeliui, o sudėtingas – galingesniam.
