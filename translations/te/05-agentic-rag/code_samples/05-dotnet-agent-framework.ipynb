# 🔍 Microsoft Foundry (.NET) తో ఎంటర్ప్రైజ్ RAG

## 📋 నేర్చుకునే లక్ష్యాలు

ఈ నోుట్‌బుక్ Microsoft Foundry తో .NET లో Microsoft Agent Framework ఉపయోగించి ఎంటერპ్రైజ్-గేస్ Retrieval-Augmented Generation (RAG) సిస్టమ్స్ ఎలా నిర్మించాలో చూపిస్తుంది. మీరు పత్రాల ద్వారా శోధించి, ఖచ్చితమైన, సందర్భ-సూచీ ప్రతిస్పందనలను ఎంటర్ప్రైజ్ భద్రత మరియు స్కేలబిలిటీతో అందించే ఉత్పత్తి-సిద్ధమైన ఏజెంట్లను తయారుచేయడం నేర్పుకుంటారు.

**మీరు నిర్మించబోయే ఎంటర్ప్రైజ్ RAG సామర్థ్యాలు:**
- 📚 **డాక్యుమెంట్ ఇంటెలిజెన్స్**: Azure AI సేవలతో అధునాతన పత్ర ప్రాసెసింగ్
- 🔍 **సెమాంటిక్ సెర్చ్**: ఎంటర్ప్రైజ్ లక్షణాలతో అధిక ప్రదర్శన వెక్టర్ శోధన
- 🛡️ **భద్రత సమగ్రత**: పాత్ర ఆధారిత ప్రవేశం మరియు డేటా రక్షణ నమూనాలు
- 🏢 **స్కేలబుల్ ఆర్కిటెక్చర్**: మనిటరింగ్‌తో ప్రొడక్షన్-రెడీ RAG సిస్టమ్స్

## 🎯 ఎంటర్ప్రైజ్ RAG ఆర్కిటెక్చర్

### ప్రధాన ఎంటర్ప్రైజ్ భాగాలు
- **Microsoft Foundry**: భద్రత మరియు అనుగుణతతో నిర్వహించే ఎంటర్ప్రైజ్ AI ప్లాట్‌ఫాం
- **ప్నిర్వహణ ఏజెంట్లు**: సంభాషణ చరిత్ర మరియు సందర్భ నిర్వహణతో స్థితిగత ఏజెంట్లు
- **వెక్టర్ స్టోర్ నిర్వహణ**: ఎంటర్ప్రైజ్-గ్రేడ్ డాక్యుమెంట్ సూచిక మరియు పొందిక
- **గుర్తింపు సమగ్రత**: Azure AD ప్రామాణీకరణ మరియు పాత్ర ఆధారిత ప్రవేశ నియంత్రణ

### .NET ఎంటర్ప్రైజ్ లాభాలు
- **టైప్ సేఫ్టీ**: RAG ఆపరేషన్‌లు మరియు డేటా నిర్మాణాల కోసం కంపైల్-టైమ్ ధ్రువీకరణ
- **అసింక్ పనితనం**: నిరోధన లేకుండా పత్ర ప్రాసెసింగ్ మరియు శోధన ఆపరేషన్లు
- **మెమరీ నిర్వహణ**: పెద్ద పత్ర సేకరణలకు సమర్ధవంతమైన వనరుల వినియోగం
- **సమగ్రత నమూనాలు**: డిపెండెన్సీ ఇంజెక్షన్‌తో స్థానిక Azure సేవ సమగ్రత

## 🏗️ సాంకేతిక ఆర్కిటెక్చర్

### ఎంటర్ప్రైజ్ RAG పైప్లైన్
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### ప్రధాన .NET భాగాలు
- **Azure.AI.Agents.Persistent**: స్థితిగత స్థితితో ఎంటర్ప్రైజ్ ఏజెంట్ నిర్వహణ
- **Azure.Identity**: భద్రత గల Azure సేవ ప్రవేశానికి సమగ్ర ప్రామాణీకరణ
- **Microsoft.Agents.AI.AzureAI**: Azure-ఆప్టిమైజ్డ్ ఏజెంట్ ఫ్రేమ్‌వర్క్ అమలు
- **System.Linq.Async**: అధిక ప్రదర్శన అసింక్ LINQ ఆపరేషన్లు

- **పాత్ర ఆధారిత ప్రవేశం**: పత్ర ప్రవేశం మరియు ఆపరేషన్‌లకు సన్నని అనుమతులు
- **డేటా రక్షణ**: అతిక్షుణమైన పత్రాలకు విశ్రాంతి మరియు ప్రయాణంలో ఎन्क్రిప్షన్
- **ఆడిట్ లాగింగ్**: అనుగుణత అవసరాలకు విస్తృత కార్యకలాప ట్రాకింగ్

### పనితనం & స్కేలబిలిటీ
- **కనెక్షన్ పూలింగ్**: సమర్ధవంతమైన Azure సేవ కనెక్షన్ నిర్వహణ
- **అసింక్ ప్రాసెసింగ్**: అధిక థ్రూపుట్ పరిస్థితుల కోసం నిరోధన లేని ఆపరేషన్లు
- **కాచింగ్ వ్యూహాలు**: తరచుగా యాక్సెస్ చేసే పత్రాలకు తెలివైన కాచింగ్
- **లోడ్ బాలెన్సింగ్**: పెద్ద-స్థాయి అమరికలకు పంపిణీ ప్రాసెసింగ్

### నిర్వహణ & మానిటరింగ్
- **హెల్త్ చెక్‌లు**: RAG సిస్టమ్ భాగాల కోసం బిల్ట్-ఇన్ మానిటరింగ్
- **పనితనం మెట్రిక్స్**: శోధన నాణ్యత మరియు ప్రతిస్పందన సమయాల వివరణాత్మక విశ్లేషణలు
- **పొరపాటు నిర్వహణ**: రీట్రై పాలసీలతో విస్తృత ఎక్సెప్షన్ నిర్వహణ
- **కాన్ఫిగరేషన్ నిర్వహణ**: ధ్రువీకరణతో పర్యావరణ-ప్రత్యేక సెట్టింగులు

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-5-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
