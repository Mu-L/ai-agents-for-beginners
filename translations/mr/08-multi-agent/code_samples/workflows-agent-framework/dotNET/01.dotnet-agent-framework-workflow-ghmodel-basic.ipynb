# 🔄 Azure OpenAI (Responses API) सह मूलभूत एजंट वर्कफ्लो (.NET)

## 📋 वर्कफ्लो ऑर्केस्ट्रेशन ट्यूटोरियल

हा नोटबुक Microsoft Agent Framework for .NET आणि Azure OpenAI (Responses API) वापरून प्रगत **एजंट वर्कफ्लो** कसे तयार करायचे हे दर्शवितो. तुम्ही बहु-टप्प्याचे व्यावसायिक प्रक्रिया तयार करायला शिकाल जिथे AI एजंट संरचित ऑर्केस्ट्रेशन नमुन्यांद्वारे जटिल कार्य पूर्ण करण्यासाठी सहकार्य करतात.

## 🎯 शिकण्याची उद्दिष्टे

### 🏗️ **वर्कफ्लो आर्किटेक्चर मूलतत्त्वे**
- **वर्कफ्लो बिल्डर**: जटिल बहु-टप्प्याच्या AI प्रक्रियांचे डिझाइन आणि ऑर्केस्ट्रेशन करा
- **एजंट समन्वय**: वर्कफ्लोमध्ये अनेक विशेषीकृत एजंटांचे समन्वय करा
- **Azure OpenAI (Responses API)**: वर्कफ्लोमध्ये Azure OpenAI Responses API चा वापर करा
- **दृश्य वर्कफ्लो डिझाइन**: चांगल्या समजासाठी वर्कफ्लो स्ट्रक्चर्स तयार आणि दृश्यमान करा

### 🔄 **प्रक्रिया ऑर्केस्ट्रेशन नमुने**
- **सकल प्रक्रिया**: लॉजिकल क्रमाने अनेक एजंट कार्यांची साखळी तयार करा
- **स्थिती व्यवस्थापन**: वर्कफ्लो टप्प्यांमध्ये संदर्भ आणि डाटा फ्लो कायम ठेवा
- **त्रुटी हाताळणी**: मजबुत त्रुटी पुनर्प्राप्ती आणि वर्कफ्लो चपळतेची अंमलबजावणी करा
- **कार्यक्षमता सुधारणा**: एंटरप्राइझ स्तरावरच्या ऑपरेशन्ससाठी कार्यक्षम वर्कफ्लो डिझाइन करा

### 🏢 **एंटरप्राइझ वर्कफ्लो अनुप्रयोग**
- **व्यवसाय प्रक्रिया ऑटोमेशन**: जटिल संघटनात्मक वर्कफ्लोज ऑटोमेट करा
- **सामग्री उत्पादन पाइपलाइन**: पुनरावलोकन आणि मान्यता टप्प्यांसह संपादकीय वर्कफ्लोज
- **ग्राहक सेवा ऑटोमेशन**: बहु-टप्प्याचा ग्राहक चौकशी निराकरण
- **डेटा प्रक्रिया वर्कफ्लो**: AI-चालित रूपांतरणांसह ETL वर्कफ्लोज

### 🔑 **Azure OpenAI कॉन्फिगरेशन**

**पर्यावरण सेटअप (.env फाइल):**
```env
AZURE_OPENAI_ENDPOINT=https://<your-resource>.openai.azure.com
AZURE_OPENAI_DEPLOYMENT=gpt-4.1-mini
```

**Azure OpenAI प्रवेश:**
1. Azure पोर्टलमध्ये Azure OpenAI संसाधन तयार करा
2. एक मॉडेल (`gpt-4.1-mini` उदाहरणार्थ) तैनात करा आणि तैनात नाव लक्षात ठेवा
3. `az login` वापरून साइन इन करा आणि वरीलप्रमाणे पर्यावरणातील चल सेट करा

### 🏗️ **वर्कफ्लो आर्किटेक्चर आढावा**

```mermaid
graph TD
    A[वर्कफ्लो बिल्डर] --> B[एजंट नोंदणी]
    B --> C[वर्कफ्लो अंमलबजावणी इंजिन]
    C --> D[Agent 1: सामग्री निर्माणकर्ता]
    C --> E[Agent 2: सामग्री पुनरावलोकनकर्ता] 
    D --> F[वर्कफ्लो निकाल]
    E --> F
    G[अॅझ्युर ओपेनएआय (प्रतिसाद API)] --> D
    G --> E
```

**महत्त्वपूर्ण घटक:**
- **WorkflowBuilder**: वर्कफ्लो डिझाइनसाठी मुख्य ऑर्केस्ट्रेशन इंजिन
- **AIAgent**: विशिष्ट क्षमता असलेले वैयक्तिकृत विशेष एजंट
- **Azure OpenAI Client**: Azure OpenAI Responses API चे एकत्रीकरण
- **Execution Context**: वर्कफ्लो टप्प्यांमध्ये स्थिती आणि डेटा फ्लो व्यवस्थापित करते

## 🎨 **एंटरप्राइझ वर्कफ्लो डिझाइन नमुने**

```
User Request → Content Generation → Quality Review → Final Output
```

```

```

### 🤝 **ग्राहक सेवा ऑटोमेशन**
```
Customer Inquiry → Classification → Processing → Response Generation → Follow-up
```

- **संसाधन व्यवस्थापन**: AI मॉडेल संसाधनांचे कार्यक्षम वाटप आणि वापर

- **डेटा गोपनीयता**: वर्कफ्लो दरम्यान संवेदनशील माहितीची सुरक्षित हाताळणी

- **त्रुटी अहवाल**: तपशीलवार त्रुटी विश्लेषण आणि डीबगिंग क्षमता
- **कार्यक्षमता विश्लेषण**: सुधारणा आणि क्षमता नियोजनासाठी मेट्रिक्स

````polyglot-notebook
using System;
using System.ComponentModel;
using Azure.AI.OpenAI;
using Azure.Identity;
using Microsoft.Extensions.AI;
using Microsoft.Agents.AI;
using Microsoft.Agents.AI.Workflows;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
// Azure OpenAI with the Responses API (stable v1 endpoint). Sign in with `az login`.
var azureEndpoint = Environment.GetEnvironmentVariable("AZURE_OPENAI_ENDPOINT") ?? throw new InvalidOperationException("AZURE_OPENAI_ENDPOINT is not set.");
var deployment = Environment.GetEnvironmentVariable("AZURE_OPENAI_DEPLOYMENT") ?? "gpt-4.1-mini";
````

````polyglot-notebook
// The Azure OpenAI client is created directly from the endpoint and Azure CLI credential — no custom client options are required.
````

````polyglot-notebook
var azureClient = new AzureOpenAIClient(new Uri(azureEndpoint), new AzureCliCredential());
````

````polyglot-notebook
const string ReviewerAgentName = "Concierge";
const string ReviewerAgentInstructions = @"
    You are an are hotel concierge who has opinions about providing the most local and authentic experiences for travelers.
    The goal is to determine if the front desk travel agent has recommended the best non-touristy experience for a traveler.
    If so, state that it is approved.
    If not, provide insight on how to refine the recommendation without using a specific example. ";
````

````polyglot-notebook
const string FrontDeskAgentName = "FrontDesk";
const string FrontDeskAgentInstructions = @"""
    You are a Front Desk Travel Agent with ten years of experience and are known for brevity as you deal with many customers.
    The goal is to provide the best activities and locations for a traveler to visit.
    Only provide a single recommendation per response.
    You're laser focused on the goal at hand.
    Don't waste time with chit chat.
    Consider suggestions when refining an idea.
    """;
````

````polyglot-notebook
AIAgent reviewerAgent = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:ReviewerAgentName,instructions:ReviewerAgentInstructions);
AIAgent frontDeskAgent  = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:FrontDeskAgentName,instructions:FrontDeskAgentInstructions);
````

````polyglot-notebook
var workflow = new WorkflowBuilder(frontDeskAgent)
            .AddEdge(frontDeskAgent, reviewerAgent)
            .Build();
````

````polyglot-notebook
ChatMessage userMessage = new ChatMessage(ChatRole.User, [
	new TextContent("I would like to go to Paris.") 
]);
````

````polyglot-notebook
StreamingRun run = await InProcessExecution.StreamAsync(workflow, userMessage);
````

````polyglot-notebook
await run.TrySendMessageAsync(new TurnToken(emitEvents: true));
string id="";
string messageData="";
await foreach (WorkflowEvent evt in run.WatchStreamAsync().ConfigureAwait(false))
{
    if (evt is AgentRunUpdateEvent executorComplete)
    {
        if(id=="")
        {
            id=executorComplete.ExecutorId;
        }
        if(id==executorComplete.ExecutorId)
        {
            messageData+=executorComplete.Data.ToString();
        }
        else
        {
            id=executorComplete.ExecutorId;
        }
        // Console.WriteLine($"{executorComplete.ExecutorId}: {executorComplete.Data}");
    }
}

Console.WriteLine(messageData);
````
