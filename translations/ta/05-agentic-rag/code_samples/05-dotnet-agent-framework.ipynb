# 🔍 Microsoft Foundry (.NET) உடன் எண்டர்பிரைஸ் RAG

## 📋 கற்றல் குறிக்கோள்கள்

இந்த நோட்புக் Microsoft Foundry உடன் .NET இல் Microsoft Agent Framework பயன்படுத்தி எண்டர்பிரைஸ் தரமான Retrieval-Augmented Generation (RAG) அமைப்புகளை எப்படி உருவாக்குவது என்பதை காட்டுகிறது. ஆவணங்களில் தேடும்படி மற்றும் எண்டர்பிரைஸ் பாதுகாப்பு மற்றும் அளவுகோலுடன் துல்லியமான, சூழல்நிலை அறிந்த பதில்களை வழங்கக்கூடிய உற்பத்தி தயார் முகவர்கள் உருவாக்க கற்றுக் கொள்வீர்கள்.

**நீங்கள் உருவாக்கப்போகும் எண்டர்பிரைஸ் RAG திறன்கள்:**
- 📚 **ஆவண் நுண்ணறிவு**: Azure AI சேவைகள் மூலம் முன்னேற்றப்படுத்தப்பட்ட ஆவண் செயலாக்கம்
- 🔍 **சமஸ்டிகத் தேடல்**: எண்டர்பிரைஸ் அம்சங்களுடன் உயர் செயல்திறன் வெக்டர் தேடல்
- 🛡️ **பாதுகாப்பு ஒருங்கிணைப்பு**: பாத்திர அடிப்படையிலான அணுகல் மற்றும் தரவுப் பாதுகாப்பு மாதிரிகள்
- 🏢 **அளவீட்டுக் கட்டமைப்பு**: கண்காணிப்பு உடன் தயாரிப்பு ரக கூட்டமைப்புகள்

## 🎯 எண்டர்பிரைஸ் RAG கட்டமைப்பு

### முக்கிய எண்டர்பிரைஸ் கூறுகள்
- **Microsoft Foundry**: பாதுகாப்பு மற்றும் ஒத்துழைப்பு கொண்ட பராமரிக்கப்பட்ட எண்டர்பிரைஸ் AI மேடையம்
- **திடமான முகவர்கள்**: உரையாடல் வரலாறு மற்றும் சூழல் மேலாண்மையுடன் நிலையான முகவர்கள்
- **வெக்டர் சேமிப்பு மேலாண்மை**: எண்டர்பிரைஸ் தரமான ஆவண பட்டியலிடல் மற்றும் மீட்பு
- **அடையாள ஒருங்கிணைப்பு**: Azure AD அங்கீகாரம் மற்றும் பாத்திர அடிப்படையிலான அணுகல் கட்டுப்பாடு

### .NET எண்டர்பிரைஸ் நன்மைகள்
- **வகை பாதுகாப்பு**: RAG செயல்பாடுகள் மற்றும் தரவு அமைப்புகளுக்கான தொகுப்புக் கால சரிபார்ப்பு
- **அசிங்க் செயல்திறன்**: தடைகள் இல்லாத ஆவண் செயலாக்கம் மற்றும் தேடல் செயல்பாடுகள்
- **மெமரி மேலாண்மை**: பெரிய ஆவணச் சேகரங்களுக்கு திறம்பட வளங்களைக் பயன்படுத்தல்
- **ஒருங்கிணைப்பு மாதிரிகள்**: சார்பு ஊற்றுடன் உள்ளூர் Azure சேவை ஒருங்கிணைப்பு

## 🏗️ தொழில்நுட்ப கட்டமைப்பு

### எண்டர்பிரைஸ் RAG குழாய்
```csharp
Document Upload → Security Validation → Vector Processing → Index Creation
                      ↓                    ↓                  ↓
User Query → Authentication → Semantic Search → Context Ranking → AI Response
```

### முக்கிய .NET கூறுகள்
- **Azure.AI.Agents.Persistent**: நிலையான முகவர் மேலாண்மை
- **Azure.Identity**: பாதுகாப்பான Azure சேவை அணுகலுக்கான ஒருங்கிணைந்த அங்கீகாரம்
- **Microsoft.Agents.AI.AzureAI**: Azure ஆற்றியுள்ள முகவர் கட்டமைப்பு செயலாக்கம்
- **System.Linq.Async**: உயர் செயல்திறன் அசிங்க் LINQ செயல்பாடுகள்

- **சேமிப்பு முறைகள்**: அடிக்கடி அணுகப்படும் ஆவணங்களுக்கான அறிவுமிக்க கேச்சிங்

## 📊 எண்டர்பிரைஸ் RAG மாதிரிகள்

### ஆவணி மேலாண்மை மாதிரிகள்
- **பெரு பதிவேற்று**: பெரிய ஆவணச் சேகரங்களின் திறன்மிக்க செயலாக்கம்
- **அதிரடி புதுப்பிப்புகள்**: நேரடி ஆவண் சேர்க்கை மற்றும் திருத்தம்
- **பதிப்பு கட்டுப்பாடு**: ஆவணம் பதிப்பு மற்றும் மாற்றங்கள் கண்காணிப்பு
- **மெட்டாடேட்டா மேலாண்மை**: தனியான ஆவண பண்புகள் மற்றும் வகைப்பாடு

### தேடல் மற்றும் மீட்பு மாதிரிகள்
- **இணை தேடல்**: சிறந்த முடிவுகளுக்காக சமஸ்டிகத் மற்றும் முக்கியச்சொல் தேடல்களின் இணைவு
- **பல்கொள்கை தேடல்**: பன்முக வடிகட்டி மற்றும் வகைப்பாடு
- **பொருத்தத்தை சரிசெய்தல்**: துறை சார்ந்த தேவைகளுக்கான தனிப்பயன் மதிப்பீட்டு அல்காரிதங்கள்
- **முடிவு தரவரிசை**: வணிக தர்க்க ஒருங்கிணைக்கப்பட்ட முன்னேற்றப்பட்ட தரவரிசை

### பாதுகாப்பு மாதிரிகள்
- **ஆவணம் மட்டுமின்றி பாதுகாப்பு**: ஆவண பகுதிக்கான நுட்ப அணுகல் கட்டுப்பாடு
- **தரவு வகைப்படுத்தல்**: தானாகவே உணர்திறன் அடையாளம் காணலும் பாதுகாப்பும்
- **ஆடிட் பாதைகள்**: அனைத்து RAG செயல்பாடுகளின் விரிவான பதிவேடு
- **குட்டீயம் பாதுகாப்பு**: PII கண்டறிதல் மற்றும் மறைத்தல் திறன்கள்

- **அதிரடி கேள்வி பதில்கள் அமைப்புகள்**: ஆவண சேகரங்களிலிருந்து இயக்கக்கூடிய FAQ உருவாக்கம்

### விதி ஒத்துழைப்பு
- **சட்ட ஆவண பகுப்பாய்வு**: ஒப்பந்த மற்றும் சட்ட ஆவண நுண்ணறிவு

- **ஆரோக்கிய டாஷ்போர்டுகள்**: நேரடி அமைப்பு ஆரோக்கிய மற்றும் செயல்திறன் காட்சி

````polyglot-notebook
using System;
using System.Linq;
using Azure.AI.Agents.Persistent;
using Azure.Identity;
using Microsoft.Agents.AI;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
var azure_foundry_endpoint = Environment.GetEnvironmentVariable("AZURE_AI_PROJECT_ENDPOINT") ?? throw new InvalidOperationException("AZURE_AI_PROJECT_ENDPOINT is not set.");
var azure_foundry_model_id = Environment.GetEnvironmentVariable("AZURE_AI_MODEL_DEPLOYMENT_NAME") ?? "gpt-5-mini";
````

````polyglot-notebook
string pdfPath = "./document.md";
````

````polyglot-notebook
using System.IO;

async Task<Stream> OpenImageStreamAsync(string path)
{
	return await Task.Run(() => File.OpenRead(path));
}

var pdfStream = await OpenImageStreamAsync(pdfPath);
````

````polyglot-notebook
var persistentAgentsClient = new PersistentAgentsClient(azure_foundry_endpoint, new AzureCliCredential());
````

````polyglot-notebook
PersistentAgentFileInfo fileInfo = await persistentAgentsClient.Files.UploadFileAsync(pdfStream, PersistentAgentFilePurpose.Agents, "demo.md");
````

````polyglot-notebook
PersistentAgentsVectorStore fileStore =
            await persistentAgentsClient.VectorStores.CreateVectorStoreAsync(
                [fileInfo.Id],
                metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
PersistentAgent agentModel = await persistentAgentsClient.Administration.CreateAgentAsync(
            azure_foundry_model_id,
            name: "DotNetRAGAgent",
            tools: [new FileSearchToolDefinition()],
            instructions: """
                You are an AI assistant designed to answer user questions using only the information retrieved from the provided document(s).

                - If a user's question cannot be answered using the retrieved context, **you must clearly respond**: 
                "I'm sorry, but the uploaded document does not contain the necessary information to answer that question."
                - Do not answer from general knowledge or reasoning. Do not make assumptions or generate hypothetical explanations.
                - Do not provide definitions, tutorials, or commentary that is not explicitly grounded in the content of the uploaded file(s).
                - If a user asks a question like "What is a Neural Network?", and this is not discussed in the uploaded document, respond as instructed above.
                - For questions that do have relevant content in the document (e.g., Contoso's travel insurance coverage), respond accurately, and cite the document explicitly.

                You must behave as if you have no external knowledge beyond what is retrieved from the uploaded document.
                """,
            toolResources: new()
            {
                FileSearch = new()
                {
                    VectorStoreIds = { fileStore.Id },
                }
            },
            metadata: new Dictionary<string, string>() { { "agentkey", bool.TrueString } });
````

````polyglot-notebook
AIAgent agent = await persistentAgentsClient.GetAIAgentAsync(agentModel.Id);
````

````polyglot-notebook
AgentThread thread = agent.GetNewThread();
````

````polyglot-notebook
Console.WriteLine(await agent.RunAsync("Can you explain Contoso's travel insurance coverage?", thread));
````
