# 🔄 Azure OpenAI (Responses API) सँग आधारभूत एजेन्ट वर्कफ्लोजहरू (.NET)

## 📋 वर्कफ्लो अर्गनाइजेशन ट्युटोरियल

यो नोटबुकले Microsoft Agent Framework for .NET र Azure OpenAI (Responses API) प्रयोग गरी परिष्कृत **एजेन्ट वर्कफ्लोज** कसरी बनाउने देखाउँछ। तपाईंले बहु-चरणिय व्यावसायिक प्रक्रियाहरू सिर्जना गर्न सिक्नुहुनेछ जहाँ AI एजेन्टहरूले संरचित अर्गनाइजेशन ढाँचाहरू मार्फत जटिल कार्यहरू पूरा गर्न सहकार्य गर्छन्।

## 🎯 सिकाइ लक्ष्यहरू

### 🏗️ **वर्कफ्लो वास्तुकला आधारहरू**
- **वर्कफ्लो बिल्डर**: जटिल बहु-चरणिय AI प्रक्रियाहरू डिजाइन र अर्गनाइज गर्नुहोस्
- **एजेन्ट समन्वय**: वर्कफ्लोभित्र धेरै विशिष्ट एजेन्टहरूको समन्वय गर्नुहोस्
- **Azure OpenAI (Responses API)**: वर्कफ्लोहरूमा Azure OpenAI Responses API को उपयोग गर्नुहोस्
- **भिजुअल वर्कफ्लो डिजाइन**: राम्रो बुझाइका लागि वर्कफ्लो संरचनाहरू सिर्जना र दृश्यात्मक गर्नुहोस्

### 🔄 **प्रक्रिया अर्गनाइजेशन ढाँचाहरू**
- **क्रमिक प्रक्रिया**: धेरै एजेन्ट कार्यहरूलाई तार्किक क्रममा जोड्नुहोस्
- **स्थिति व्यवस्थापन**: वर्कफ्लो चरणहरूमा सन्दर्भ र डाटा प्रवाह कायम राख्नुहोस्
- **त्रुटि ह्यान्डलिङ**: दृष्‍टिगोचर त्रुटि पुन:प्राप्ति र वर्कफ्लो स्थायित्व लागू गर्नुहोस्
- **प्रदर्शन अनुकूलन**: उद्यम-स्तरका अपरेसनहरूका लागि कुशल वर्कफ्लो डिजाइन गर्नुहोस्

### 🏢 **उद्यम वर्कफ्लो अनुप्रयोगहरू**
- **व्यापार प्रक्रिया स्वचालन**: जटिल संगठनात्मक वर्कफ्लोहरू स्वचालित बनाउनुहोस्
- **सामग्री उत्पादन पाइपलाइन**: समीक्षा र स्वीकृति चरणहरू सहितको सम्पादकीय वर्कफ्लोज
- **ग्राहक सेवा स्वचालन**: बहु-चरण ग्राहक सोधपुछ समाधान
- **डाटा प्रक्रिया वर्कफ्लोज**: AI-समर्थित रूपान्तरणसहित ETL वर्कफ्लोज

### 🔑 **Azure OpenAI कन्फिगरेसन**

**Environment Setup (.env फाइल):**
```env
AZURE_OPENAI_ENDPOINT=https://<your-resource>.openai.azure.com
AZURE_OPENAI_DEPLOYMENT=gpt-4.1-mini
```

**Azure OpenAI पहुँच:**
1. Azure पोर्टलमा Azure OpenAI स्रोत बनाउनुहोस्
2. एउटा मोडेल (जस्तै, `gpt-4.1-mini`) तैनाथ गर्नुहोस् र तैनाथी नाम नोट गर्नुहोस्
3. `az login` सँग साइन इन गर्नुहोस् र माथि देखाइएको अनुसार वातावरण भेरिएबलहरू कन्फिगर गर्नुहोस्

### 🏗️ **वर्कफ्लो वास्तुकला अवलोकन**

```mermaid
graph TD
    A[कार्यप्रवाह निर्माता] --> B[एजेन्ट दर्ता]
    B --> C[कार्यप्रवाह कार्यान्वयन इञ्जिन]
    C --> D[Agent 1: सामग्री उत्पादक]
    C --> E[Agent 2: सामग्री समीक्षक] 
    D --> F[कार्यप्रवाह परिणामहरू]
    E --> F

    G --> E
```

**मुख्य कम्पोनेंटहरू:**
- **WorkflowBuilder**: वर्कफ्लो डिजाइन गर्न मुख्य अर्गनाइजेशन इञ्जिन
- **AIAgent**: विशेष क्षमताहरू सहित व्यक्तिगत विशिष्ट एजेन्टहरू
- **Azure OpenAI Client**: Azure OpenAI Responses API एकीकरण
- **Execution Context**: वर्कफ्लो चरणहरूबीच स्थिति र डाटा प्रवाह व्यवस्थापन गर्छ

````polyglot-notebook
using System;
using System.ComponentModel;
using Azure.AI.OpenAI;
using Azure.Identity;
using Microsoft.Extensions.AI;
using Microsoft.Agents.AI;
using Microsoft.Agents.AI.Workflows;
````

````polyglot-notebook
 using DotNetEnv;
````

````polyglot-notebook
Env.Load("../../../.env");
````

````polyglot-notebook
// Azure OpenAI with the Responses API (stable v1 endpoint). Sign in with `az login`.
var azureEndpoint = Environment.GetEnvironmentVariable("AZURE_OPENAI_ENDPOINT") ?? throw new InvalidOperationException("AZURE_OPENAI_ENDPOINT is not set.");
var deployment = Environment.GetEnvironmentVariable("AZURE_OPENAI_DEPLOYMENT") ?? "gpt-4.1-mini";
````

````polyglot-notebook
// The Azure OpenAI client is created directly from the endpoint and Azure CLI credential — no custom client options are required.
````

````polyglot-notebook
var azureClient = new AzureOpenAIClient(new Uri(azureEndpoint), new AzureCliCredential());
````

````polyglot-notebook
const string ReviewerAgentName = "Concierge";
const string ReviewerAgentInstructions = @"
    You are an are hotel concierge who has opinions about providing the most local and authentic experiences for travelers.
    The goal is to determine if the front desk travel agent has recommended the best non-touristy experience for a traveler.
    If so, state that it is approved.
    If not, provide insight on how to refine the recommendation without using a specific example. ";
````

````polyglot-notebook
const string FrontDeskAgentName = "FrontDesk";
const string FrontDeskAgentInstructions = @"""
    You are a Front Desk Travel Agent with ten years of experience and are known for brevity as you deal with many customers.
    The goal is to provide the best activities and locations for a traveler to visit.
    Only provide a single recommendation per response.
    You're laser focused on the goal at hand.
    Don't waste time with chit chat.
    Consider suggestions when refining an idea.
    """;
````

````polyglot-notebook
AIAgent reviewerAgent = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:ReviewerAgentName,instructions:ReviewerAgentInstructions);
AIAgent frontDeskAgent  = azureClient.GetOpenAIResponseClient(deployment).CreateAIAgent(
    name:FrontDeskAgentName,instructions:FrontDeskAgentInstructions);
````

````polyglot-notebook
var workflow = new WorkflowBuilder(frontDeskAgent)
            .AddEdge(frontDeskAgent, reviewerAgent)
            .Build();
````

````polyglot-notebook
ChatMessage userMessage = new ChatMessage(ChatRole.User, [
	new TextContent("I would like to go to Paris.") 
]);
````

````polyglot-notebook
StreamingRun run = await InProcessExecution.StreamAsync(workflow, userMessage);
````

````polyglot-notebook
await run.TrySendMessageAsync(new TurnToken(emitEvents: true));
string id="";
string messageData="";
await foreach (WorkflowEvent evt in run.WatchStreamAsync().ConfigureAwait(false))
{
    if (evt is AgentRunUpdateEvent executorComplete)
    {
        if(id=="")
        {
            id=executorComplete.ExecutorId;
        }
        if(id==executorComplete.ExecutorId)
        {
            messageData+=executorComplete.Data.ToString();
        }
        else
        {
            id=executorComplete.ExecutorId;
        }
        // Console.WriteLine($"{executorComplete.ExecutorId}: {executorComplete.Data}");
    }
}

Console.WriteLine(messageData);
````
